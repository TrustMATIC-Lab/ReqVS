```python
import sqlite3
import pandas as pd
import os
from tqdm import tqdm

def export_all_tables_to_csv(db_path, output_dir="csv_exports"):
    """Export all tables from SQLite database to CSV files"""
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"Created output directory: {output_dir}")
    
    print(f"Connecting to database: {db_path}")
    conn = sqlite3.connect(db_path)
    
    cursor = conn.cursor()
    cursor.execute("""
        SELECT name FROM sqlite_master 
        WHERE type='table' 
        AND name NOT LIKE 'sqlite_%'
        ORDER BY name;
    """)
    
    tables = [row[0] for row in cursor.fetchall()]
    print(f"Found {len(tables)} tables\n")
    
    success_count = 0
    error_count = 0
    total_rows = 0
    
    for table_name in tqdm(tables, desc="Exporting tables"):
        try:
            query = f"SELECT * FROM {table_name};"
            df = pd.read_sql(query, conn)
            
            output_file = os.path.join(output_dir, f"{table_name}.csv")
            df.to_csv(output_file, index=False, encoding='utf-8')
            
            success_count += 1
            total_rows += len(df)
            
            print(f"✓ {table_name}: {len(df):,} rows, {len(df.columns)} columns -> {output_file}")
            
        except Exception as e:
            error_count += 1
            print(f"✗ {table_name}: Export failed - {str(e)}")
    
    conn.close()
    
    print("\n" + "="*60)
    print("Export completed!")
    print(f"Success: {success_count} tables")
    print(f"Failed: {error_count} tables")
    print(f"Total rows: {total_rows:,}")
    print(f"Output directory: {os.path.abspath(output_dir)}")
    print("="*60)

if __name__ == "__main__":
    db_path = "chembl_36.db"
    
    if not os.path.exists(db_path):
        print(f"Error: Database file not found: {db_path}")
        print("Please ensure you are running this script in the directory containing chembl_36.db")
        exit(1)
    
    export_all_tables_to_csv(db_path, output_dir="csv_exports")
```

```python

"""
Merge activities.csv with SMILES and sequence data
Only keep entries with protein targets and small molecules (with SMILES)
"""

import csv
import re
from collections import defaultdict
from pathlib import Path
import os

PARENT_DIR = Path.cwd()
BASE_DIR = os.path.join(PARENT_DIR, "csv_exports")

ACTIVITIES_FILE = os.path.join(BASE_DIR, "activities.csv")
COMPOUND_STRUCTURES_FILE = os.path.join(BASE_DIR, "compound_structures.csv")
ASSAYS_FILE = os.path.join(BASE_DIR, "assays.csv")
TARGET_DICTIONARY_FILE = os.path.join(BASE_DIR, "target_dictionary.csv")
TARGET_COMPONENTS_FILE = os.path.join(BASE_DIR, "target_components.csv")
COMPONENT_SEQUENCES_FILE = os.path.join(BASE_DIR, "component_sequences.csv")
FASTA_FILE = os.path.join(PARENT_DIR, "chembl_36.fa")
OUTPUT_FILE = os.path.join(PARENT_DIR, "step2_activities_with_smiles_sequence.csv")

PROTEIN_TARGET_TYPES = {
    'SINGLE PROTEIN',
}

print("Starting processing...")
print(f"Protein target types: {', '.join(sorted(PROTEIN_TARGET_TYPES))}\n")

print("Loading SMILES data...")
smiles_dict = {}
with open(COMPOUND_STRUCTURES_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        molregno = row.get('molregno', '').strip()
        canonical_smiles = row.get('canonical_smiles', '').strip()
        if molregno and canonical_smiles:
            smiles_dict[molregno] = canonical_smiles
print(f"Loaded {len(smiles_dict)} SMILES records")

print("Loading assays data...")
assay_to_tid = {}
with open(ASSAYS_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        assay_id = row.get('assay_id', '').strip()
        tid = row.get('tid', '').strip()
        if assay_id and tid:
            assay_to_tid[assay_id] = tid
print(f"Loaded {len(assay_to_tid)} assay records")

print("Loading target_dictionary data...")
tid_to_target_type = {}
tid_to_chembl_id = {}
protein_tids = set()
with open(TARGET_DICTIONARY_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        tid = row.get('tid', '').strip()
        target_type = row.get('target_type', '').strip()
        chembl_id = row.get('chembl_id', '').strip()
        
        if tid:
            tid_to_target_type[tid] = target_type
            if chembl_id:
                tid_to_chembl_id[tid] = chembl_id
            
            if target_type in PROTEIN_TARGET_TYPES:
                protein_tids.add(tid)

print(f"Loaded {len(tid_to_target_type)} target records")
print(f"Protein targets: {len(protein_tids)}")

print("Loading target_components data...")
tid_to_components = defaultdict(list)
with open(TARGET_COMPONENTS_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        tid = row.get('tid', '').strip()
        component_id = row.get('component_id', '').strip()
        if tid and component_id:
            tid_to_components[tid].append(component_id)
print(f"Loaded {len(tid_to_components)} target-component association records")

print("Loading FASTA file data...")
chembl_id_to_sequence = {}
if os.path.exists(FASTA_FILE):
    current_chembl_id = None
    current_sequence = []
    
    with open(FASTA_FILE, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if line.startswith('>'):
                if current_chembl_id and current_sequence:
                    chembl_id_to_sequence[current_chembl_id] = ''.join(current_sequence)
                
                match = re.match(r'>\s*(CHEMBL\d+)', line)
                if match:
                    current_chembl_id = match.group(1)
                    current_sequence = []
                else:
                    current_chembl_id = None
            elif current_chembl_id and line:
                current_sequence.append(line)
        
        if current_chembl_id and current_sequence:
            chembl_id_to_sequence[current_chembl_id] = ''.join(current_sequence)
    
    print(f"Loaded {len(chembl_id_to_sequence)} FASTA sequence records")
else:
    print(f"Warning: FASTA file does not exist: {FASTA_FILE}")

print("Loading component_sequences data...")
component_to_sequence = {}
with open(COMPONENT_SEQUENCES_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        component_id = row.get('component_id', '').strip()
        sequence = row.get('sequence', '').strip()
        if component_id and sequence:
            component_to_sequence[component_id] = sequence
print(f"Loaded {len(component_to_sequence)} component sequence records")

print("\nProcessing activities.csv...")
print("Filter: Only keep protein targets + SMILES + sequences\n")

processed_count = 0
skipped_non_protein = 0
skipped_no_smiles = 0
skipped_no_sequence = 0
written_count = 0

with open(ACTIVITIES_FILE, 'r', encoding='utf-8') as infile, \
     open(OUTPUT_FILE, 'w', encoding='utf-8', newline='') as outfile:
    
    reader = csv.DictReader(infile)
    
    fieldnames = list(reader.fieldnames)
    fieldnames.append('canonical_smiles')
    fieldnames.append('sequence')
    
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()
    
    for row in reader:
        processed_count += 1
        if processed_count % 1000000 == 0:
            print(f"Processed {processed_count:,} rows, written {written_count:,} rows...")
        
        assay_id = row.get('assay_id', '').strip()
        if not assay_id:
            continue
        
        tid = assay_to_tid.get(assay_id, '')
        if not tid or tid not in protein_tids:
            skipped_non_protein += 1
            continue
        
        molregno = row.get('molregno', '').strip()
        canonical_smiles = smiles_dict.get(molregno, '')
        if not canonical_smiles:
            skipped_no_smiles += 1
            continue
        
        sequence = ''
        chembl_id = tid_to_chembl_id.get(tid, '')
        if chembl_id:
            sequence = chembl_id_to_sequence.get(chembl_id, '')
        
        if not sequence:
            component_ids = tid_to_components.get(tid, [])
            for comp_id in component_ids:
                seq = component_to_sequence.get(comp_id, '')
                if seq:
                    sequence = seq
                    break
        
        if not sequence:
            skipped_no_sequence += 1
            continue
        
        row['canonical_smiles'] = canonical_smiles
        row['sequence'] = sequence
        writer.writerow(row)
        written_count += 1

print(f"\nProcessing completed!")
print(f"=" * 60)
print(f"Total processed rows: {processed_count:,}")
print(f"Written to output file: {written_count:,} ({written_count/processed_count*100:.2f}%)")
print(f"\nFilter statistics:")
print(f"  - Skipped non-protein targets: {skipped_non_protein:,} ({skipped_non_protein/processed_count*100:.2f}%)")
print(f"  - Skipped no SMILES: {skipped_no_smiles:,} ({skipped_no_smiles/processed_count*100:.2f}%)")
print(f"  - Skipped no sequence: {skipped_no_sequence:,} ({skipped_no_sequence/processed_count*100:.2f}%)")
print(f"\nOutput file: {OUTPUT_FILE}")
print(f"Output file contains: {written_count:,} rows (protein targets + SMILES + sequences)")
```

```python

"""
Clean activities_with_smiles_sequence.csv
- Keep only confidence score 8/9
- Remove molecules with >70 heavy atoms
- Remove proteins with >2000 residues
- Keep only relation='=', types: Ki, Kd, IC50, EC50
- Keep units=nM, add neg_log10_value column
"""

import csv
import math
from pathlib import Path

try:
    from rdkit import Chem
    RDKIT_AVAILABLE = True
except ImportError:
    RDKIT_AVAILABLE = False
    print("Warning: RDKit not installed, using simplified heavy atom counting")

BASE_DIR = Path.cwd()
INPUT_FILE = BASE_DIR / "step2_activities_with_smiles_sequence.csv"
ASSAYS_FILE = BASE_DIR / "csv_exports/assays.csv"
OUTPUT_FILE = BASE_DIR / "step3_activities_cleaned.csv"

ALLOWED_TYPES = {'Ki', 'Kd', 'IC50', 'EC50'}

def count_heavy_atoms_simple(smiles):
    """Simplified heavy atom counting (without RDKit)"""
    if not smiles:
        return 0
    
    import re
    smiles_clean = re.sub(r'\[[^\]]+\]', '', smiles)
    
    heavy_atoms = 0
    i = 0
    while i < len(smiles_clean):
        char = smiles_clean[i]
        if char in '()[]=+-.0123456789@#$%':
            i += 1
            continue
        
        if char.isupper():
            if i + 1 < len(smiles_clean) and smiles_clean[i+1].islower():
                heavy_atoms += 1
                i += 2
            else:
                if char != 'H':
                    heavy_atoms += 1
                i += 1
        else:
            i += 1
    
    return heavy_atoms

def count_heavy_atoms_rdkit(smiles):
    """Calculate heavy atom count using RDKit"""
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return 0
        return mol.GetNumHeavyAtoms()
    except:
        return 0

def count_heavy_atoms(smiles):
    """Calculate heavy atom count for SMILES"""
    if not smiles:
        return 0
    
    if RDKIT_AVAILABLE:
        return count_heavy_atoms_rdkit(smiles)
    else:
        return count_heavy_atoms_simple(smiles)

def count_residues(sequence):
    """Calculate residue count for protein sequence"""
    if not sequence:
        return 0
    return len(sequence.strip())

def calculate_neg_log10_from_nM(value_nM):
    """
    Calculate negative log10: convert nM to M then take negative log10
    Formula: neg_log10 = -log10(value_nM × 10^-9) = 9 - log10(value_nM)
    """
    try:
        value_float = float(value_nM)
        if value_float <= 0:
            return None
        
        value_M = value_float * 1e-9
        neg_log10 = -math.log10(value_M)
        
        return neg_log10
    except (ValueError, TypeError):
        return None

print("Starting data cleaning...\n")

print("1. Loading assays data (get confidence scores)...")
assay_to_confidence = {}
with open(ASSAYS_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        assay_id = row.get('assay_id', '').strip()
        confidence_score = row.get('confidence_score', '').strip()
        if assay_id and confidence_score:
            try:
                confidence_score = int(confidence_score)
                assay_to_confidence[assay_id] = confidence_score
            except ValueError:
                pass
print(f"   Loaded {len(assay_to_confidence)} assay confidence records")

print("\n2. Processing activities data...")
print("   Filter conditions:")
print("   - Confidence = 8 or 9")
print("   - Relation = '='")
print("   - Types: Ki, Kd, IC50, EC50")
print("   - Units = nM")
print("   - Heavy atoms <= 70")
print("   - Residues <= 2000")
print("   - Negative log10 = -log10(value_nM × 10^-9) = 9 - log10(value_nM)\n")

stats = {
    'total': 0,
    'skipped_no_confidence': 0,
    'skipped_confidence_not_8_9': 0,
    'skipped_relation_not_equal': 0,
    'skipped_type_not_allowed': 0,
    'skipped_units_not_nm': 0,
    'skipped_heavy_atoms_over_70': 0,
    'skipped_residues_over_2000': 0,
    'skipped_invalid_value': 0,
    'written': 0
}

with open(INPUT_FILE, 'r', encoding='utf-8') as infile, \
     open(OUTPUT_FILE, 'w', encoding='utf-8', newline='') as outfile:
    
    reader = csv.DictReader(infile)
    
    fieldnames = list(reader.fieldnames)
    if 'neg_log10_value' not in fieldnames:
        fieldnames.append('neg_log10_value')
    
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()
    
    for row in reader:
        stats['total'] += 1
        if stats['total'] % 100000 == 0:
            print(f"   Processed {stats['total']:,} rows, written {stats['written']:,} rows...")
        
        assay_id = row.get('assay_id', '').strip()
        if not assay_id:
            stats['skipped_no_confidence'] += 1
            continue
        
        confidence_score = assay_to_confidence.get(assay_id)
        if confidence_score is None:
            stats['skipped_no_confidence'] += 1
            continue
        
        if confidence_score != 9 and confidence_score != 8:
            stats['skipped_confidence_not_8_9'] += 1
            continue
        
        standard_relation = row.get('standard_relation', '').strip()
        if standard_relation != '=':
            stats['skipped_relation_not_equal'] += 1
            continue
        
        standard_type = row.get('standard_type', '').strip()
        if standard_type not in ALLOWED_TYPES:
            stats['skipped_type_not_allowed'] += 1
            continue
        
        standard_units = row.get('standard_units', '').strip()
        if standard_units.upper() != 'NM':
            stats['skipped_units_not_nm'] += 1
            continue
        
        canonical_smiles = row.get('canonical_smiles', '').strip()
        if canonical_smiles:
            heavy_atoms = count_heavy_atoms(canonical_smiles)
            if heavy_atoms > 70:
                stats['skipped_heavy_atoms_over_70'] += 1
                continue
        else:
            stats['skipped_heavy_atoms_over_70'] += 1
            continue
        
        sequence = row.get('sequence', '').strip()
        if sequence:
            residues = count_residues(sequence)
            if residues > 2000:
                stats['skipped_residues_over_2000'] += 1
                continue
        else:
            stats['skipped_residues_over_2000'] += 1
            continue
        
        standard_value = row.get('standard_value', '').strip()
        neg_log10 = calculate_neg_log10_from_nM(standard_value)
        if neg_log10 is None:
            stats['skipped_invalid_value'] += 1
            continue
        
        row['neg_log10_value'] = f"{neg_log10:.4f}"
        
        writer.writerow(row)
        stats['written'] += 1

print(f"\nProcessing completed!")
print(f"=" * 70)
print(f"Total processed rows: {stats['total']:,}")
print(f"Written to output file: {stats['written']:,} ({stats['written']/stats['total']*100:.2f}%)")
print(f"\nFilter statistics:")
print(f"  - Skipped no confidence: {stats['skipped_no_confidence']:,} ({stats['skipped_no_confidence']/stats['total']*100:.2f}%)")
print(f"  - Skipped confidence not 8/9: {stats['skipped_confidence_not_8_9']:,} ({stats['skipped_confidence_not_8_9']/stats['total']*100:.2f}%)")
print(f"  - Skipped relation not '=': {stats['skipped_relation_not_equal']:,} ({stats['skipped_relation_not_equal']/stats['total']*100:.2f}%)")
print(f"  - Skipped type not allowed: {stats['skipped_type_not_allowed']:,} ({stats['skipped_type_not_allowed']/stats['total']*100:.2f}%)")
print(f"  - Skipped units not nM: {stats['skipped_units_not_nm']:,} ({stats['skipped_units_not_nm']/stats['total']*100:.2f}%)")
print(f"  - Skipped heavy atoms >70: {stats['skipped_heavy_atoms_over_70']:,} ({stats['skipped_heavy_atoms_over_70']/stats['total']*100:.2f}%)")
print(f"  - Skipped residues >2000: {stats['skipped_residues_over_2000']:,} ({stats['skipped_residues_over_2000']/stats['total']*100:.2f}%)")
print(f"  - Skipped invalid value: {stats['skipped_invalid_value']:,} ({stats['skipped_invalid_value']/stats['total']*100:.2f}%)")
print(f"\nOutput file: {OUTPUT_FILE}")
print(f"\nFormula: neg_log10 = -log10(value_nM × 10^-9) = 9 - log10(value_nM)")
print(f"Example: 1000 nM -> -log10(1000 × 10^-9) = -log10(10^-6) = 6.0")
```

```python

"""
Create classification dataset from activities_cleaned.csv
- Keep only neg_log10_value > 7 or < 5
- neg_log10_value < 5: Y = 0 (inactive)
- neg_log10_value > 7: Y = 1 (active)
- Exclude 5 <= neg_log10_value <= 7
- For duplicate (protein, molecule) pairs:
  * If labels consistent, keep one
  * If labels inconsistent, remove all
"""

import csv
from pathlib import Path
from collections import defaultdict

BASE_DIR = Path.cwd()
INPUT_FILE = BASE_DIR / "step3_activities_cleaned.csv"
OUTPUT_FILE = BASE_DIR / "step4_classification_dataset.csv"

print("Starting classification dataset creation...\n")
print("Filter conditions:")
print("  - neg_log10_value > 7: Y = 1 (active)")
print("  - neg_log10_value < 5: Y = 0 (inactive)")
print("  - Exclude 5 <= neg_log10_value <= 7")
print("  - Duplicate handling: keep one if consistent, remove all if inconsistent\n")

stats = {
    'total': 0,
    'active': 0,
    'inactive': 0,
    'excluded': 0,
    'invalid': 0,
    'duplicate_pairs': 0,
    'duplicate_consistent': 0,
    'duplicate_inconsistent': 0,
    'duplicate_removed': 0
}

print("Step 1: Collect and classify data...")
pair_records = defaultdict(list)

with open(INPUT_FILE, 'r', encoding='utf-8') as infile:
    reader = csv.DictReader(infile)
    
    for row in reader:
        stats['total'] += 1
        if stats['total'] % 100000 == 0:
            print(f"  Processed {stats['total']:,} rows, active: {stats['active']:,}, inactive: {stats['inactive']:,}...")
        
        sequence = row.get('sequence', '').strip()
        canonical_smiles = row.get('canonical_smiles', '').strip()
        standard_type = row.get('standard_type', '').strip()
        standard_value = row.get('standard_value', '').strip()
        standard_units = row.get('standard_units', '').strip()
        neg_log10_value_str = row.get('neg_log10_value', '').strip()
        
        if not sequence or not canonical_smiles or not neg_log10_value_str:
            stats['invalid'] += 1
            continue
        
        try:
            neg_log10_value = float(neg_log10_value_str)
        except (ValueError, TypeError):
            stats['invalid'] += 1
            continue
        
        if neg_log10_value > 7:
            Y = 1
            stats['active'] += 1
        elif neg_log10_value < 5:
            Y = 0
            stats['inactive'] += 1
        else:
            stats['excluded'] += 1
            continue
        
        record = {
            'Protein': sequence,
            'SMILES': canonical_smiles,
            'Y': Y,
            'type': standard_type,
            'neg_log10_value': f"{neg_log10_value:.4f}",
            'standard_value': standard_value,
            'standard_units': standard_units
        }
        
        key = (sequence, canonical_smiles)
        pair_records[key].append(record)

print(f"\nStep 2: Handle duplicate data...")
print(f"  Unique (Protein, SMILES) pairs: {len(pair_records):,}")

records_to_write = []

for key, records in pair_records.items():
    if len(records) > 1:
        stats['duplicate_pairs'] += 1
        
        y_values = [r['Y'] for r in records]
        unique_y_values = set(y_values)
        
        if len(unique_y_values) == 1:
            stats['duplicate_consistent'] += 1
            records_to_write.append(records[0])
            stats['duplicate_removed'] += len(records) - 1
        else:
            stats['duplicate_inconsistent'] += 1
            stats['duplicate_removed'] += len(records)
    else:
        records_to_write.append(records[0])

print(f"  Duplicate pair statistics:")
print(f"    - Total duplicate pairs: {stats['duplicate_pairs']:,}")
print(f"    - Consistent labels (keep one): {stats['duplicate_consistent']:,}")
print(f"    - Inconsistent labels (remove all): {stats['duplicate_inconsistent']:,}")
print(f"    - Records removed due to duplicates: {stats['duplicate_removed']:,}")

print(f"\nStep 3: Write output file...")
output_fields = [
    'Protein',
    'SMILES',
    'Y',
    'type',
    'neg_log10_value',
    'standard_value',
    'standard_units'
]

with open(OUTPUT_FILE, 'w', encoding='utf-8', newline='') as outfile:
    writer = csv.DictWriter(outfile, fieldnames=output_fields)
    writer.writeheader()
    
    for i, record in enumerate(records_to_write):
        writer.writerow(record)
        if (i + 1) % 100000 == 0:
            print(f"  Written {i + 1:,} records...")

print(f"\nProcessing completed!")
print(f"=" * 70)
print(f"Total processed rows: {stats['total']:,}")
print(f"\nClassification statistics:")
print(f"  - Active (Y=1, neg_log10_value > 7): {stats['active']:,} ({stats['active']/stats['total']*100:.2f}%)")
print(f"  - Inactive (Y=0, neg_log10_value < 5): {stats['inactive']:,} ({stats['inactive']/stats['total']*100:.2f}%)")
print(f"  - Excluded (5 <= neg_log10_value <= 7): {stats['excluded']:,} ({stats['excluded']/stats['total']*100:.2f}%)")
print(f"  - Invalid data: {stats['invalid']:,} ({stats['invalid']/stats['total']*100:.2f}%)")
print(f"\nDuplicate handling statistics:")
print(f"  - Duplicate (Protein, SMILES) pairs: {stats['duplicate_pairs']:,}")
print(f"  - Consistent duplicates (keep one): {stats['duplicate_consistent']:,}")
print(f"  - Inconsistent duplicates (remove all): {stats['duplicate_inconsistent']:,}")
print(f"  - Records removed: {stats['duplicate_removed']:,}")
print(f"\nOutput dataset:")
initial_count = stats['active'] + stats['inactive']
final_count = len(records_to_write)
print(f"  - Samples before deduplication: {initial_count:,}")
print(f"  - Samples after deduplication: {final_count:,}")
print(f"  - Removed samples: {initial_count - final_count:,} ({((initial_count - final_count)/initial_count*100) if initial_count > 0 else 0:.2f}%)")
print(f"\nOutput file: {OUTPUT_FILE}")
```

```python

"""
Cluster protein sequences in step5_classification_dataset_cleaned.csv
Using cd-hit tool, 70% similarity
Add target_cluster column to each entry
"""

import csv
import subprocess
import re
from pathlib import Path

BASE_DIR = Path.cwd()
INPUT_FILE = BASE_DIR / "step5_classification_dataset_cleaned.csv"
OUTPUT_FILE = BASE_DIR / "step7_classification_dataset_with_clusters.csv"

FASTA_FILE = BASE_DIR / "proteins_for_clustering.fasta"
CDHIT_OUTPUT = BASE_DIR / "proteins_clustered"
CDHIT_CLSTR = BASE_DIR / "proteins_clustered.clstr"

print("=" * 80)
print("Protein Sequence Clustering")
print("=" * 80)
print()

print("1. Extracting all unique protein sequences...")
unique_proteins = {}
protein_index = 0

with open(INPUT_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        protein = row.get('Protein', '').strip()
        if protein and protein not in unique_proteins:
            unique_proteins[protein] = protein_index
            protein_index += 1

print(f"   Found {len(unique_proteins):,} unique protein sequences")
print()

print("2. Writing sequences to FASTA file...")
with open(FASTA_FILE, 'w', encoding='utf-8') as f:
    for protein, idx in unique_proteins.items():
        f.write(f">seq_{idx}\n")
        for i in range(0, len(protein), 80):
            f.write(protein[i:i+80] + "\n")

print(f"   Written {len(unique_proteins):,} sequences to {FASTA_FILE}")
print()

print("3. Running cd-hit for clustering (70% similarity)...")
print("   Command: cd-hit -i proteins_for_clustering.fasta -o proteins_clustered -c 0.7 -n 5 -T 0 -M 0")
print()

try:
    result = subprocess.run(['cd-hit', '-h'], capture_output=True, text=True, timeout=5)
    cdhit_available = True
except (FileNotFoundError, subprocess.TimeoutExpired):
    cdhit_available = False
    print("   Error: cd-hit not found, please ensure cd-hit is installed and in PATH")
    print("   Installation: conda install -c bioconda cd-hit")
    exit(1)

cmd = [
    'cd-hit',
    '-i', str(FASTA_FILE),
    '-o', str(CDHIT_OUTPUT),
    '-c', '0.7',
    '-n', '5',
    '-T', '0',
    '-M', '0'
]

print("   Running cd-hit (this may take some time)...")
try:
    result = subprocess.run(cmd, capture_output=True, text=True, check=True)
    print("   cd-hit completed")
    if result.stderr:
        lines = result.stderr.split('\n')
        for line in lines:
            if any(keyword in line.lower() for keyword in ['total seq', 'clusters', 'longest', 'shortest']):
                print(f"   {line.strip()}")
except subprocess.CalledProcessError as e:
    print(f"   Error: cd-hit failed")
    print(f"   Error message: {e.stderr}")
    exit(1)

print()

print("4. Parsing clustering results...")

if not CDHIT_CLSTR.exists():
    print(f"   Error: Clustering result file does not exist: {CDHIT_CLSTR}")
    exit(1)

sequence_to_cluster = {}
current_cluster = None
cluster_count = 0

with open(CDHIT_CLSTR, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line.startswith('>Cluster'):
            match = re.search(r'>Cluster\s+(\d+)', line)
            if match:
                current_cluster = int(match.group(1))
                cluster_count = max(cluster_count, current_cluster + 1)
        elif line and current_cluster is not None:
            match = re.search(r'>seq_(\d+)', line)
            if match:
                seq_idx = int(match.group(1))
                sequence_to_cluster[seq_idx] = current_cluster

print(f"   Clustered into {cluster_count:,} clusters")
print(f"   Parsed {len(sequence_to_cluster):,} sequence cluster assignments")
print()

print("5. Building protein sequence to cluster mapping...")
protein_to_cluster = {}

index_to_protein = {idx: seq for seq, idx in unique_proteins.items()}

for seq_idx, cluster_id in sequence_to_cluster.items():
    if seq_idx in index_to_protein:
        protein = index_to_protein[seq_idx]
        protein_to_cluster[protein] = cluster_id

unassigned_count = 0
for protein in unique_proteins:
    if protein not in protein_to_cluster:
        protein_to_cluster[protein] = cluster_count
        cluster_count += 1
        unassigned_count += 1

if unassigned_count > 0:
    print(f"   Warning: {unassigned_count} sequences not found in clustering results, assigned new clusters")

print(f"   Built {len(protein_to_cluster):,} sequence mappings")
print()

print("6. Adding target_cluster column to original data...")

stats = {
    'total': 0,
    'with_cluster': 0,
    'without_cluster': 0
}

with open(INPUT_FILE, 'r', encoding='utf-8') as infile, \
     open(OUTPUT_FILE, 'w', encoding='utf-8', newline='') as outfile:
    
    reader = csv.DictReader(infile)
    fieldnames = list(reader.fieldnames)
    
    if 'target_cluster' in fieldnames:
        fieldnames.remove('target_cluster')
    
    fieldnames.append('target_cluster')
    
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()
    
    for row in reader:
        stats['total'] += 1
        
        protein = row.get('Protein', '').strip()
        cluster_id = protein_to_cluster.get(protein)
        
        if cluster_id is not None:
            row['target_cluster'] = cluster_id
            stats['with_cluster'] += 1
        else:
            row['target_cluster'] = cluster_count
            stats['without_cluster'] += 1
        
        writer.writerow(row)
        
        if stats['total'] % 100000 == 0:
            print(f"   Processed {stats['total']:,} rows...")

print()

print("7. Statistics on cluster distribution...")
cluster_sizes = {}
with open(OUTPUT_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        cluster_id = row.get('target_cluster', '').strip()
        if cluster_id:
            try:
                cluster_id = int(cluster_id)
                cluster_sizes[cluster_id] = cluster_sizes.get(cluster_id, 0) + 1
            except ValueError:
                pass

print(f"   Number of clusters: {len(cluster_sizes):,}")
if cluster_sizes:
    sizes = sorted(cluster_sizes.values())
    print(f"   Minimum cluster size: {min(sizes):,}")
    print(f"   Maximum cluster size: {max(sizes):,}")
    print(f"   Average cluster size: {sum(sizes)/len(sizes):.2f}")
    print(f"   Median cluster size: {sizes[len(sizes)//2]:,}")

size_distribution = {}
for size in sizes:
    if size < 10:
        size_distribution['1-9'] = size_distribution.get('1-9', 0) + 1
    elif size < 100:
        size_distribution['10-99'] = size_distribution.get('10-99', 0) + 1
    elif size < 1000:
        size_distribution['100-999'] = size_distribution.get('100-999', 0) + 1
    else:
        size_distribution['1000+'] = size_distribution.get('1000+', 0) + 1

print(f"\n   Cluster size distribution:")
for size_range, count in sorted(size_distribution.items()):
    print(f"     - {size_range}: {count:,} clusters")

print()

print("=" * 80)
print("Clustering completed!")
print("=" * 80)
print(f"Input file: {INPUT_FILE}")
print(f"Output file: {OUTPUT_FILE}")
print(f"FASTA file: {FASTA_FILE}")
print(f"cd-hit output: {CDHIT_OUTPUT}")
print()
print(f"Statistics:")
print(f"  - Unique protein sequences: {len(unique_proteins):,}")
print(f"  - Clusters after clustering: {cluster_count:,}")
print(f"  - Total data records: {stats['total']:,}")
print(f"  - Records with clusters: {stats['with_cluster']:,}")
if stats['without_cluster'] > 0:
    print(f"  - Records without clusters: {stats['without_cluster']:,}")
print()
print(f"Clustering parameters: 70% similarity")
print("=" * 80)
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from rdkit import Chem
from rdkit import DataStructs
from rdkit.Chem import rdFingerprintGenerator
import argparse
import os
from typing import Optional, Dict, List, Tuple
from pathlib import Path

MORGAN_GENERATOR = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048, includeChirality=False)

def calculate_fingerprint(smiles: str) -> Optional[object]:
    """
    Calculate 2048-bit binary fingerprint using Morgan algorithm (radius=2)
    
    Corresponds to paper:
    "For each compound, we computed a 2048-bit binary Morgan fingerprint using RDKit."
    """
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None
        fingerprint = MORGAN_GENERATOR.GetFingerprint(mol)
        return fingerprint
    except Exception as e:
        print(f"[FP ERROR] SMILES = {smiles}, error = {e}")
        return None


def precompute_fingerprints(df: pd.DataFrame, smiles_col: str = "SMILES") -> List[Optional[object]]:
    """
    Precompute fingerprints for all rows in df, returns a list aligned with df.index
    """
    fps: List[Optional[object]] = []
    print("Precomputing Morgan fingerprints for all compounds...")
    for smi in tqdm(df[smiles_col].tolist(), desc="Fingerprints"):
        fps.append(calculate_fingerprint(smi))
    valid = sum(fp is not None for fp in fps)
    print(f"  - Total compounds: {len(fps)}, valid fingerprints: {valid}")
    return fps


def build_cluster_index(df_positive: pd.DataFrame) -> Dict[int, List[int]]:
    """
    Build {cluster_id: [row_index_list]} mapping based on target_cluster.
    Uses integer index from df_positive (after reset_index).
    """
    cluster_ids = df_positive["target_cluster"].values
    cluster_index: Dict[int, List[int]] = {}
    for idx, c in enumerate(cluster_ids):
        cluster_index.setdefault(c, []).append(idx)
    return cluster_index


def is_candidate_dissimilar_to_cluster(
    cand_fp: object,
    cluster_pos_indices: List[int],
    fps_positive: List[Optional[object]],
    threshold: float
) -> bool:
    """
    Check if candidate ligand fingerprint satisfies Tanimoto(candidate, each_positive) < threshold
    for all positive samples in the cluster.
    
    Corresponds to paper:
    "A candidate was accepted as negative only if its Tanimoto similarity
    to all of the current positive samples was less than 0.3"
    """
    if cand_fp is None:
        return False

    for idx in cluster_pos_indices:
        fp_pos = fps_positive[idx]
        if fp_pos is None:
            continue
        sim = DataStructs.TanimotoSimilarity(cand_fp, fp_pos)
        if sim >= threshold:
            return False
    return True


def generate_negatives_for_cluster(
    cluster_id: int,
    cluster_pos_indices: List[int],
    df_positive: pd.DataFrame,
    fps_positive: List[Optional[object]],
    global_used_candidate_indices: set,
    similarity_threshold: float,
    max_trials_factor: float = 10.0
) -> List[Tuple[int, int]]:
    """
    Generate negative samples for a cluster. For each positive sample in the cluster,
    select a candidate ligand from other clusters, requiring:
      - Candidate belongs to other clusters
      - Candidate not used before (globally unique)
      - Candidate Tanimoto similarity to all positives in this cluster < threshold
    Returns list of [(pos_idx, cand_idx), ...], length ideally equals cluster positive count.
    
    Corresponds to paper:
      - "iteratively handled each cluster"
      - "treating all positive samples within the cluster as active"
      - "selected an equal number of candidate ligands from other clusters"
      - "each negative sample was used only once"
    """
    n_pos = len(cluster_pos_indices)
    if n_pos == 0:
        return []

    all_indices = np.arange(len(df_positive))
    cluster_arr = df_positive["target_cluster"].values

    candidate_mask = (cluster_arr != cluster_id)
    candidate_indices = all_indices[candidate_mask]

    candidate_indices = [
        idx for idx in candidate_indices
        if (idx not in global_used_candidate_indices) and (fps_positive[idx] is not None)
    ]

    if len(candidate_indices) == 0:
        print(f"[WARN] Cluster {cluster_id}: no available candidate ligands from other clusters.")
        return []

    rng = np.random.default_rng()
    rng.shuffle(candidate_indices)

    max_trials = int(max_trials_factor * n_pos)
    if max_trials > len(candidate_indices):
        max_trials = len(candidate_indices)

    selected_pairs: List[Tuple[int, int]] = []
    used_in_this_cluster: set = set()

    cluster_pos_indices_local = list(cluster_pos_indices)

    cand_ptr = 0
    good_candidates: List[int] = []

    while len(good_candidates) < n_pos and cand_ptr < max_trials:
        cand_idx = candidate_indices[cand_ptr]
        cand_ptr += 1

        if cand_idx in global_used_candidate_indices or cand_idx in used_in_this_cluster:
            continue

        cand_fp = fps_positive[cand_idx]
        if cand_fp is None:
            continue

        if is_candidate_dissimilar_to_cluster(
            cand_fp,
            cluster_pos_indices_local,
            fps_positive,
            similarity_threshold
        ):
            good_candidates.append(cand_idx)
            used_in_this_cluster.add(cand_idx)
            global_used_candidate_indices.add(cand_idx)

    if len(good_candidates) == 0:
        print(f"[WARN] Cluster {cluster_id}: no dissimilar candidates found (under threshold={similarity_threshold}).")
        return []

    if len(good_candidates) < n_pos:
        print(
            f"[WARN] Cluster {cluster_id}: only found {len(good_candidates)} dissimilar candidates "
            f"for {n_pos} positives. Will only augment the first {len(good_candidates)} positives."
        )

    rng.shuffle(cluster_pos_indices_local)
    pair_count = min(len(cluster_pos_indices_local), len(good_candidates))
    for i in range(pair_count):
        pos_idx = cluster_pos_indices_local[i]
        cand_idx = good_candidates[i]
        selected_pairs.append((pos_idx, cand_idx))

    return selected_pairs


def augment_data_according_to_paper(
    input_file: str,
    output_file: str,
    similarity_threshold: float = 0.3,
    random_seed: int = 0
) -> None:
    """
    Implement data augmentation according to paper section 2.1.2:
      - Based on target_cluster (from CD-HIT clustering)
      - Process at cluster level
      - All positives in a cluster treated as actives
      - Negatives from ligands in other clusters
      - Candidate must have Tanimoto similarity < threshold to all positives in cluster
      - Each candidate used only once
      - Aim for #positives == #negatives per cluster (may be less if insufficient candidates)
    
    Generated negative sample rows:
      - Copy a positive row from the cluster
      - Keep Protein/target_cluster target information
      - Replace SMILES with ligand from other cluster
      - Set Y=0
    """
    np.random.seed(random_seed)

    print(f"Reading data from {input_file} ...")
    df = pd.read_csv(input_file)

    if "Y" not in df.columns:
        raise ValueError("Input CSV must contain column 'Y' as label (1 for positive, 0 for negative).")
    if "SMILES" not in df.columns:
        raise ValueError("Input CSV must contain column 'SMILES'.")
    if "target_cluster" not in df.columns:
        raise ValueError("Input CSV must contain column 'target_cluster'.")
    if "Protein" not in df.columns:
        raise ValueError("Input CSV must contain column 'Protein'.")

    df_positive = df[df["Y"] == 1].copy()
    df_positive.reset_index(drop=True, inplace=True)

    print(f"Total positives: {len(df_positive)}")

    fps_positive = precompute_fingerprints(df_positive, smiles_col="SMILES")

    cluster_index = build_cluster_index(df_positive)
    all_clusters = sorted(cluster_index.keys())
    print(f"Number of target clusters (with positives): {len(all_clusters)}")

    global_used_candidate_indices: set = set()

    negative_rows = []

    for cluster_id in tqdm(all_clusters, desc="Processing clusters"):
        pos_indices = cluster_index[cluster_id]
        if len(pos_indices) == 0:
            continue

        pairs = generate_negatives_for_cluster(
            cluster_id=cluster_id,
            cluster_pos_indices=pos_indices,
            df_positive=df_positive,
            fps_positive=fps_positive,
            global_used_candidate_indices=global_used_candidate_indices,
            similarity_threshold=similarity_threshold,
            max_trials_factor=10.0,
        )

        for pos_idx, cand_idx in pairs:
            pos_row = df_positive.loc[pos_idx]
            cand_row = df_positive.loc[cand_idx]

            new_row = pos_row.copy()
            new_row["SMILES"] = cand_row["SMILES"]
            new_row["Y"] = 0

            negative_rows.append(new_row)

    if len(negative_rows) == 0:
        print("[WARN] No negative samples were generated. Please check your threshold or data.")
        df_aug = df
    else:
        df_negative = pd.DataFrame(negative_rows)
        print(f"Generated negative samples: {len(df_negative)}")

        df_aug = pd.concat([df, df_negative], ignore_index=True)

    print(f"Saving augmented dataset to {output_file} ...")
    df_aug.to_csv(output_file, index=False)
    print("Done.")
    print(f"Original samples: {len(df)}")
    print(f"Augmented samples: {len(df_aug)}")


def main():
    parser = argparse.ArgumentParser(
        description="Data augmentation according to the paper (cluster-wise conditional label inversion)."
    )
    parser.add_argument("--input", "-i", required=True, help="Input CSV file path")
    parser.add_argument("--output", "-o", required=True, help="Output CSV file path")
    parser.add_argument(
        "--threshold", "-t", type=float, default=0.3,
        help="Tanimoto similarity threshold for negative sampling (default: 0.3)"
    )
    parser.add_argument("--seed", "-s", type=int, default=0, help="Random seed (default: 0)")

    args = parser.parse_args()

    if not os.path.exists(args.input):
        print(f"Error: input file {args.input} not found!")
        return

    out_dir = os.path.dirname(args.output)
    if out_dir and not os.path.exists(out_dir):
        os.makedirs(out_dir)

    augment_data_according_to_paper(
        input_file=args.input,
        output_file=args.output,
        similarity_threshold=args.threshold,
        random_seed=args.seed
    )


if __name__ == "__main__":
    BASE_DIR = Path.cwd()
    INPUT_FILE = BASE_DIR / "step7_classification_dataset_with_clusters.csv"
    OUTPUT_FILE = BASE_DIR / "step8_classification_dataset_augmented.csv"
    
    if not INPUT_FILE.exists():
        print(f"Error: Input file not found: {INPUT_FILE}")
        print("Please ensure you are running this script in the directory containing step7_classification_dataset_with_clusters.csv")
        exit(1)
    
    augment_data_according_to_paper(
        input_file=str(INPUT_FILE),
        output_file=str(OUTPUT_FILE),
        similarity_threshold=0.3,
        random_seed=0
    )
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Final processing step: Extract key columns and check for duplicates
- Extract from step8_classification_dataset_augmented.csv: Protein, SMILES, target_cluster, Y
- Check for duplicate (Protein, SMILES) pairs
- Check for label inconsistencies in duplicate pairs
"""

import csv
from pathlib import Path
from collections import defaultdict
from datetime import datetime

BASE_DIR = Path.cwd()
INPUT_FILE = BASE_DIR / "step8_classification_dataset_augmented.csv"
OUTPUT_FILE = BASE_DIR / "step9_final_dataset.csv"
REPORT_FILE = BASE_DIR / f"step9_duplicate_check_report_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"
DUPLICATE_INCONSISTENT_FILE = BASE_DIR / f"step9_duplicate_inconsistent_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"

print("=" * 80)
print("Final Processing: Extract Key Columns and Check Duplicates")
print("=" * 80)
print()

print("1. Reading data and extracting key columns...")

if not INPUT_FILE.exists():
    print(f"Error: File not found: {INPUT_FILE}")
    exit(1)

data_records = []
pair_records = defaultdict(list)

required_columns = ['Protein', 'SMILES', 'target_cluster', 'Y']
stats = {
    'total_rows': 0,
    'missing_columns': 0,
    'valid_rows': 0,
    'unique_pairs': 0,
    'duplicate_pairs': 0,
    'duplicate_consistent': 0,
    'duplicate_inconsistent': 0
}

with open(INPUT_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    
    missing_cols = [col for col in required_columns if col not in reader.fieldnames]
    if missing_cols:
        print(f"Error: Missing required columns: {', '.join(missing_cols)}")
        print(f"Available columns: {', '.join(reader.fieldnames)}")
        exit(1)
    
    for row in reader:
        stats['total_rows'] += 1
        
        if stats['total_rows'] % 100000 == 0:
            print(f"   Processed {stats['total_rows']:,} rows...")
        
        protein = row.get('Protein', '').strip()
        smiles = row.get('SMILES', '').strip()
        target_cluster_str = row.get('target_cluster', '').strip()
        y_str = row.get('Y', '').strip()
        
        if not protein or not smiles or not target_cluster_str or not y_str:
            stats['missing_columns'] += 1
            continue
        
        try:
            target_cluster = int(target_cluster_str)
            y = int(y_str)
        except (ValueError, TypeError):
            stats['missing_columns'] += 1
            continue
        
        record = {
            'Protein': protein,
            'SMILES': smiles,
            'target_cluster': target_cluster,
            'Y': y
        }
        
        data_records.append(record)
        stats['valid_rows'] += 1
        
        key = (protein, smiles)
        pair_records[key].append({
            'target_cluster': target_cluster,
            'Y': y,
            'row_index': len(data_records) - 1
        })

print(f"   Total rows: {stats['total_rows']:,}")
print(f"   Valid rows: {stats['valid_rows']:,}")
print(f"   Rows with missing fields: {stats['missing_columns']:,}")
print(f"   Unique (Protein, SMILES) pairs: {len(pair_records):,}")
print()

print("2. Checking for duplicates...")

duplicate_inconsistent_records = []

for key, records in pair_records.items():
    if len(records) > 1:
        stats['duplicate_pairs'] += 1
        
        y_values = [r['Y'] for r in records]
        unique_y_values = set(y_values)
        
        if len(unique_y_values) == 1:
            stats['duplicate_consistent'] += 1
        else:
            stats['duplicate_inconsistent'] += 1
            
            protein, smiles = key
            for record in records:
                duplicate_inconsistent_records.append({
                    'Protein': protein[:100] + '...' if len(protein) > 100 else protein,
                    'SMILES': smiles[:100] + '...' if len(smiles) > 100 else smiles,
                    'target_cluster': record['target_cluster'],
                    'Y': record['Y'],
                    'duplicate_count': len(records),
                    'unique_y_values': ', '.join(map(str, sorted(unique_y_values)))
                })
    else:
        stats['unique_pairs'] += 1

print(f"   Unique (Protein, SMILES) pairs: {stats['unique_pairs']:,}")
print(f"   Duplicate (Protein, SMILES) pairs: {stats['duplicate_pairs']:,}")
print(f"   Duplicates with consistent labels: {stats['duplicate_consistent']:,}")
print(f"   Duplicates with inconsistent labels: {stats['duplicate_inconsistent']:,}")
print()

print("3. Saving final dataset...")

output_columns = ['Protein', 'SMILES', 'target_cluster', 'Y']

with open(OUTPUT_FILE, 'w', encoding='utf-8', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=output_columns)
    writer.writeheader()
    
    for i, record in enumerate(data_records):
        writer.writerow(record)
        if (i + 1) % 100000 == 0:
            print(f"   Written {i + 1:,} records...")

print(f"   Saved {len(data_records):,} records to {OUTPUT_FILE}")
print()

if duplicate_inconsistent_records:
    print("4. Saving duplicate records with inconsistent labels...")
    with open(DUPLICATE_INCONSISTENT_FILE, 'w', encoding='utf-8', newline='') as f:
        fieldnames = ['Protein', 'SMILES', 'target_cluster', 'Y', 'duplicate_count', 'unique_y_values']
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(duplicate_inconsistent_records)
    print(f"   Saved {len(duplicate_inconsistent_records):,} records to {DUPLICATE_INCONSISTENT_FILE}")
    print()

print("5. Generating report...")

report_lines = []
report_lines.append("=" * 80)
report_lines.append("Final Processing Report")
report_lines.append(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
report_lines.append("=" * 80)
report_lines.append("")

report_lines.append("I. Data Extraction Statistics")
report_lines.append("-" * 80)
report_lines.append(f"Input file: {INPUT_FILE}")
report_lines.append(f"Total rows: {stats['total_rows']:,}")
report_lines.append(f"Valid rows: {stats['valid_rows']:,}")
report_lines.append(f"Rows with missing fields: {stats['missing_columns']:,}")
report_lines.append(f"Output file: {OUTPUT_FILE}")
report_lines.append(f"Output records: {len(data_records):,}")
report_lines.append("")

report_lines.append("II. Duplicate Check")
report_lines.append("-" * 80)
report_lines.append(f"Unique (Protein, SMILES) pairs: {stats['unique_pairs']:,}")
report_lines.append(f"Duplicate (Protein, SMILES) pairs: {stats['duplicate_pairs']:,}")
if stats['duplicate_pairs'] > 0:
    report_lines.append(f"  - Consistent labels: {stats['duplicate_consistent']:,} ({stats['duplicate_consistent']/stats['duplicate_pairs']*100:.2f}%)")
    report_lines.append(f"  - Inconsistent labels: {stats['duplicate_inconsistent']:,} ({stats['duplicate_inconsistent']/stats['duplicate_pairs']*100:.2f}%)")
report_lines.append("")

report_lines.append("III. Data Quality Assessment")
report_lines.append("-" * 80)
if stats['duplicate_inconsistent'] == 0:
    report_lines.append("✓ All duplicate (Protein, SMILES) pairs have consistent labels")
    report_lines.append("  Data consistency is good")
else:
    report_lines.append(f"⚠ Found {stats['duplicate_inconsistent']:,} (Protein, SMILES) pairs with inconsistent labels")
    report_lines.append(f"  Affecting {len(duplicate_inconsistent_records):,} records")
    report_lines.append(f"  Detailed records saved to: {DUPLICATE_INCONSISTENT_FILE}")
    report_lines.append("")
    report_lines.append("  Recommendations:")
    report_lines.append("  - Check the data sources of these duplicate records")
    report_lines.append("  - Consider whether deduplication is needed")
    report_lines.append("  - If labels are inconsistent, manual review or a strategy (e.g., keep majority label) may be needed")

report_lines.append("")
report_lines.append("IV. Output Files")
report_lines.append("-" * 80)
report_lines.append(f"Final dataset: {OUTPUT_FILE}")
report_lines.append("  Columns: Protein, SMILES, target_cluster, Y")
report_lines.append(f"  Records: {len(data_records):,}")
if duplicate_inconsistent_records:
    report_lines.append(f"Inconsistent label records: {DUPLICATE_INCONSISTENT_FILE}")
    report_lines.append(f"  Records: {len(duplicate_inconsistent_records):,}")

report_lines.append("")
report_lines.append("=" * 80)

with open(REPORT_FILE, 'w', encoding='utf-8') as f:
    f.write('\n'.join(report_lines))

print('\n'.join(report_lines))
print()
print(f"Report saved to: {REPORT_FILE}")

print()
print("=" * 80)
print("Processing completed!")
print("=" * 80)
```

```python

"""
Build regression training dataset from step3_activities_cleaned.csv
- Keep only Ki and Kd type entries
- Keep only samples with neg_log10_value between 0-12
- For duplicate (protein, compound) pairs:
  * If max-min difference > 1, discard
  * If difference <= 1, keep average value
- Output score distribution plot
"""

import csv
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path
from collections import defaultdict
from datetime import datetime

BASE_DIR = Path.cwd()
INPUT_FILE = BASE_DIR / "step3_activities_cleaned.csv"
OUTPUT_FILE = BASE_DIR / "step10_regression_dataset.csv"
PLOT_FILE = BASE_DIR / "step10_score_distribution.png"
REPORT_FILE = BASE_DIR / f"step10_regression_dataset_report_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"

SCORE_MIN = 0.0
SCORE_MAX = 12.0
MAX_DIFF_THRESHOLD = 1.0

print("=" * 80)
print("Building Regression Training Dataset")
print("=" * 80)
print()
print(f"Filtering conditions:")
print(f"  - Types: Ki, Kd")
print(f"  - Score range: {SCORE_MIN} <= neg_log10_value <= {SCORE_MAX}")
print(f"  - Duplicate handling: If max-min > {MAX_DIFF_THRESHOLD}, discard")
print()

print("1. Reading data and filtering...")

if not INPUT_FILE.exists():
    print(f"Error: File not found: {INPUT_FILE}")
    exit(1)

ALLOWED_TYPES = {'Ki', 'Kd'}

pair_records = defaultdict(list)
stats = {
    'total_rows': 0,
    'ki_count': 0,
    'kd_count': 0,
    'other_type_count': 0,
    'invalid_neg_log10': 0,
    'out_of_range': 0
}

with open(INPUT_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    
    required_cols = ['sequence', 'canonical_smiles', 'standard_type', 'neg_log10_value']
    missing_cols = [col for col in required_cols if col not in reader.fieldnames]
    if missing_cols:
        print(f"Error: Missing required columns: {', '.join(missing_cols)}")
        print(f"Available columns: {', '.join(reader.fieldnames)}")
        exit(1)
    
    for row in reader:
        stats['total_rows'] += 1
        
        if stats['total_rows'] % 100000 == 0:
            print(f"   Processed {stats['total_rows']:,} rows...")
        
        standard_type = row.get('standard_type', '').strip()
        if standard_type not in ALLOWED_TYPES:
            stats['other_type_count'] += 1
            continue
        
        if standard_type == 'Ki':
            stats['ki_count'] += 1
        elif standard_type == 'Kd':
            stats['kd_count'] += 1
        
        sequence = row.get('sequence', '').strip()
        canonical_smiles = row.get('canonical_smiles', '').strip()
        neg_log10_value_str = row.get('neg_log10_value', '').strip()
        
        if not sequence or not canonical_smiles or not neg_log10_value_str:
            continue
        
        try:
            neg_log10_value = float(neg_log10_value_str)
        except (ValueError, TypeError):
            stats['invalid_neg_log10'] += 1
            continue
        
        if neg_log10_value < SCORE_MIN or neg_log10_value > SCORE_MAX:
            stats['out_of_range'] += 1
            continue
        
        key = (sequence, canonical_smiles)
        pair_records[key].append({
            'sequence': sequence,
            'canonical_smiles': canonical_smiles,
            'standard_type': standard_type,
            'neg_log10_value': neg_log10_value,
            'standard_value': row.get('standard_value', '').strip(),
            'standard_units': row.get('standard_units', '').strip()
        })

print(f"   Total rows: {stats['total_rows']:,}")
print(f"   Ki type: {stats['ki_count']:,}")
print(f"   Kd type: {stats['kd_count']:,}")
print(f"   Other types: {stats['other_type_count']:,}")
print(f"   Invalid neg_log10_value: {stats['invalid_neg_log10']:,}")
print(f"   Out of range ({SCORE_MIN}-{SCORE_MAX}): {stats['out_of_range']:,}")
print(f"   Unique (Protein, SMILES) pairs: {len(pair_records):,}")
print()

print("2. Processing duplicate (Protein, SMILES) pairs...")

final_records = []
duplicate_stats = {
    'unique_pairs': 0,
    'duplicate_pairs': 0,
    'duplicate_kept': 0,
    'duplicate_dropped': 0,
    'total_duplicate_records': 0,
    'dropped_records': 0
}

for key, records in pair_records.items():
    if len(records) == 1:
        final_records.append(records[0])
        duplicate_stats['unique_pairs'] += 1
    else:
        duplicate_stats['duplicate_pairs'] += 1
        duplicate_stats['total_duplicate_records'] += len(records)
        
        neg_log10_values = [r['neg_log10_value'] for r in records]
        min_val = min(neg_log10_values)
        max_val = max(neg_log10_values)
        diff = max_val - min_val
        
        if diff > MAX_DIFF_THRESHOLD:
            duplicate_stats['duplicate_dropped'] += 1
            duplicate_stats['dropped_records'] += len(records)
        else:
            duplicate_stats['duplicate_kept'] += 1
            avg_neg_log10 = np.mean(neg_log10_values)
            
            merged_record = records[0].copy()
            merged_record['neg_log10_value'] = avg_neg_log10
            final_records.append(merged_record)

print(f"   Unique pairs: {duplicate_stats['unique_pairs']:,}")
print(f"   Duplicate pairs: {duplicate_stats['duplicate_pairs']:,}")
print(f"     - Kept (diff<={MAX_DIFF_THRESHOLD}): {duplicate_stats['duplicate_kept']:,}")
print(f"     - Dropped (diff>{MAX_DIFF_THRESHOLD}): {duplicate_stats['duplicate_dropped']:,}")
print(f"   Dropped records: {duplicate_stats['dropped_records']:,}")
print(f"   Final records: {len(final_records):,}")
print()

print("3. Saving regression dataset...")

output_columns = ['Protein', 'SMILES', 'neg_log10_value', 'type', 'standard_value', 'standard_units']

with open(OUTPUT_FILE, 'w', encoding='utf-8', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=output_columns)
    writer.writeheader()
    
    for record in final_records:
        output_row = {
            'Protein': record['sequence'],
            'SMILES': record['canonical_smiles'],
            'neg_log10_value': f"{record['neg_log10_value']:.4f}",
            'type': record['standard_type'],
            'standard_value': record['standard_value'],
            'standard_units': record['standard_units']
        }
        writer.writerow(output_row)

print(f"   Saved {len(final_records):,} records to {OUTPUT_FILE}")
print()

print("4. Plotting score distribution...")

neg_log10_values = [r['neg_log10_value'] for r in final_records]
neg_log10_array = np.array(neg_log10_values)

fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Regression Dataset Score Distribution (neg_log10_value)', fontsize=16, fontweight='bold')

ax1 = axes[0, 0]
ax1.hist(neg_log10_array, bins=50, edgecolor='black', alpha=0.7, color='steelblue')
ax1.set_xlabel('neg_log10_value', fontsize=12)
ax1.set_ylabel('Frequency', fontsize=12)
ax1.set_title('Histogram', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axvline(np.mean(neg_log10_array), color='red', linestyle='--', linewidth=2, label=f'Mean: {np.mean(neg_log10_array):.2f}')
ax1.axvline(np.median(neg_log10_array), color='green', linestyle='--', linewidth=2, label=f'Median: {np.median(neg_log10_array):.2f}')
ax1.legend()

ax2 = axes[0, 1]
ax2.hist(neg_log10_array, bins=50, density=True, edgecolor='black', alpha=0.7, color='coral')
ax2.set_xlabel('neg_log10_value', fontsize=12)
ax2.set_ylabel('Density', fontsize=12)
ax2.set_title('Density Plot', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)

ax3 = axes[1, 0]
bp = ax3.boxplot(neg_log10_array, vert=True, patch_artist=True)
bp['boxes'][0].set_facecolor('lightblue')
ax3.set_ylabel('neg_log10_value', fontsize=12)
ax3.set_title('Box Plot', fontsize=14, fontweight='bold')
ax3.grid(True, alpha=0.3)

ax4 = axes[1, 1]
ax4.axis('off')
stats_text = f"""
Statistics:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Total samples: {len(neg_log10_array):,}

Mean: {np.mean(neg_log10_array):.4f}
Median: {np.median(neg_log10_array):.4f}
Std: {np.std(neg_log10_array):.4f}

Min: {np.min(neg_log10_array):.4f}
Max: {np.max(neg_log10_array):.4f}
Q1 (25%): {np.percentile(neg_log10_array, 25):.4f}
Q3 (75%): {np.percentile(neg_log10_array, 75):.4f}

Ki samples: {sum(1 for r in final_records if r['standard_type'] == 'Ki'):,}
Kd samples: {sum(1 for r in final_records if r['standard_type'] == 'Kd'):,}

Unique pairs: {duplicate_stats['unique_pairs']:,}
Duplicate pairs: {duplicate_stats['duplicate_pairs']:,}
"""
ax4.text(0.1, 0.5, stats_text, fontsize=11, family='monospace', 
         verticalalignment='center', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.savefig(PLOT_FILE, dpi=300, bbox_inches='tight')
print(f"   Saved distribution plot to {PLOT_FILE}")
print()

print("5. Generating report...")

report_lines = []
report_lines.append("=" * 80)
report_lines.append("Regression Dataset Construction Report")
report_lines.append(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
report_lines.append("=" * 80)
report_lines.append("")

report_lines.append("I. Data Filtering Statistics")
report_lines.append("-" * 80)
report_lines.append(f"Input file: {INPUT_FILE}")
report_lines.append(f"Total rows: {stats['total_rows']:,}")
report_lines.append(f"Ki type: {stats['ki_count']:,}")
report_lines.append(f"Kd type: {stats['kd_count']:,}")
report_lines.append(f"Other types (excluded): {stats['other_type_count']:,}")
report_lines.append(f"Invalid neg_log10_value: {stats['invalid_neg_log10']:,}")
report_lines.append(f"Out of range ({SCORE_MIN}-{SCORE_MAX}): {stats['out_of_range']:,}")
report_lines.append("")

report_lines.append("II. Deduplication Statistics")
report_lines.append("-" * 80)
report_lines.append(f"Unique (Protein, SMILES) pairs: {duplicate_stats['unique_pairs']:,}")
report_lines.append(f"Duplicate (Protein, SMILES) pairs: {duplicate_stats['duplicate_pairs']:,}")
report_lines.append(f"  - Kept (diff<={MAX_DIFF_THRESHOLD}): {duplicate_stats['duplicate_kept']:,}")
report_lines.append(f"  - Dropped (diff>{MAX_DIFF_THRESHOLD}): {duplicate_stats['duplicate_dropped']:,}")
report_lines.append(f"Total duplicate records: {duplicate_stats['total_duplicate_records']:,}")
report_lines.append(f"Dropped records: {duplicate_stats['dropped_records']:,}")
report_lines.append(f"Deduplication strategy: For duplicate pairs, if max-min>{MAX_DIFF_THRESHOLD} then discard, otherwise keep average")
report_lines.append(f"Final records: {len(final_records):,}")
report_lines.append("")

report_lines.append("III. Score Distribution Statistics")
report_lines.append("-" * 80)
report_lines.append(f"Sample count: {len(neg_log10_array):,}")
report_lines.append(f"Mean: {np.mean(neg_log10_array):.4f}")
report_lines.append(f"Median: {np.median(neg_log10_array):.4f}")
report_lines.append(f"Std: {np.std(neg_log10_array):.4f}")
report_lines.append(f"Min: {np.min(neg_log10_array):.4f}")
report_lines.append(f"Max: {np.max(neg_log10_array):.4f}")
report_lines.append(f"25th percentile: {np.percentile(neg_log10_array, 25):.4f}")
report_lines.append(f"75th percentile: {np.percentile(neg_log10_array, 75):.4f}")
report_lines.append("")

report_lines.append("IV. Type Distribution")
report_lines.append("-" * 80)
ki_count = sum(1 for r in final_records if r['standard_type'] == 'Ki')
kd_count = sum(1 for r in final_records if r['standard_type'] == 'Kd')
report_lines.append(f"Ki: {ki_count:,} ({ki_count/len(final_records)*100:.2f}%)")
report_lines.append(f"Kd: {kd_count:,} ({kd_count/len(final_records)*100:.2f}%)")
report_lines.append("")

report_lines.append("V. Output Files")
report_lines.append("-" * 80)
report_lines.append(f"Dataset file: {OUTPUT_FILE}")
report_lines.append(f"  Columns: Protein, SMILES, neg_log10_value, type, standard_value, standard_units")
report_lines.append(f"  Records: {len(final_records):,}")
report_lines.append(f"Distribution plot: {PLOT_FILE}")
report_lines.append("")

report_lines.append("=" * 80)

with open(REPORT_FILE, 'w', encoding='utf-8') as f:
    f.write('\n'.join(report_lines))

print('\n'.join(report_lines))
print()
print(f"Report saved to: {REPORT_FILE}")

print()
print("=" * 80)
print("Processing completed!")
print("=" * 80)
```

```python
"""
Remove proteins and compounds overlapping with test sets (DUD-E and DEKOIS2.0) from regression dataset
Output:
1. Cleaned regression dataset (overlaps removed)
2. Data overlapping with DUD-E
3. Data overlapping with DEKOIS2.0
"""

import csv
from pathlib import Path
from datetime import datetime

BASE_DIR = Path.cwd()
REGRESSION_DATASET = BASE_DIR / "step10_regression_dataset.csv"
DUDE_FILE = BASE_DIR / "DUD-E.csv"
DEKOIS_FILE = BASE_DIR / "DEKOIS2.0.csv"

OUTPUT_CLEANED = BASE_DIR / "step11_regression_dataset_cleaned.csv"
OUTPUT_DUDE_OVERLAP = BASE_DIR / "step11_regression_dataset_dude_overlap.csv"
OUTPUT_DEKOIS_OVERLAP = BASE_DIR / "step11_regression_dataset_dekois_overlap.csv"
REPORT_FILE = BASE_DIR / f"step11_regression_test_overlap_report_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"

print("=" * 80)
print("Regression Dataset Test Set Overlap Removal")
print("=" * 80)
print()

print("1. Loading DUD-E test set...")
dude_proteins = set()
dude_smiles = set()
dude_count = 0

if DUDE_FILE.exists():
    with open(DUDE_FILE, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        for row in reader:
            dude_count += 1
            protein = row.get('Protein', '').strip()
            smiles = row.get('SMILES', '').strip()
            if protein:
                dude_proteins.add(protein)
            if smiles:
                dude_smiles.add(smiles)
    print(f"   Loaded {dude_count:,} DUD-E records")
    print(f"   - Unique proteins: {len(dude_proteins):,}")
    print(f"   - Unique SMILES: {len(dude_smiles):,}")
else:
    print(f"   Warning: DUD-E file not found: {DUDE_FILE}")

print("\n2. Loading DEKOIS2.0 test set...")
dekois_proteins = set()
dekois_smiles = set()
dekois_count = 0

if DEKOIS_FILE.exists():
    with open(DEKOIS_FILE, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        for row in reader:
            dekois_count += 1
            protein = row.get('Protein', '').strip()
            smiles = row.get('SMILES', '').strip()
            if protein:
                dekois_proteins.add(protein)
            if smiles:
                dekois_smiles.add(smiles)
    print(f"   Loaded {dekois_count:,} DEKOIS2.0 records")
    print(f"   - Unique proteins: {len(dekois_proteins):,}")
    print(f"   - Unique SMILES: {len(dekois_smiles):,}")
else:
    print(f"   Warning: DEKOIS2.0 file not found: {DEKOIS_FILE}")

all_test_proteins = dude_proteins | dekois_proteins
all_test_smiles = dude_smiles | dekois_smiles

print(f"\n3. Combined test set statistics:")
print(f"   - Total unique proteins: {len(all_test_proteins):,}")
print(f"   - Total unique SMILES: {len(all_test_smiles):,}")

print(f"\n4. Processing regression dataset...")

if not REGRESSION_DATASET.exists():
    print(f"Error: Regression dataset file not found: {REGRESSION_DATASET}")
    exit(1)

stats = {
    'total': 0,
    'dude_protein_overlap': 0,
    'dude_smiles_overlap': 0,
    'dude_both_overlap': 0,
    'dekois_protein_overlap': 0,
    'dekois_smiles_overlap': 0,
    'dekois_both_overlap': 0,
    'test_protein_overlap': 0,
    'test_smiles_overlap': 0,
    'test_both_overlap': 0,
    'cleaned': 0
}

with open(REGRESSION_DATASET, 'r', encoding='utf-8') as infile, \
     open(OUTPUT_CLEANED, 'w', encoding='utf-8', newline='') as out_cleaned, \
     open(OUTPUT_DUDE_OVERLAP, 'w', encoding='utf-8', newline='') as out_dude, \
     open(OUTPUT_DEKOIS_OVERLAP, 'w', encoding='utf-8', newline='') as out_dekois:
    
    reader = csv.DictReader(infile)
    fieldnames = reader.fieldnames
    
    writer_cleaned = csv.DictWriter(out_cleaned, fieldnames=fieldnames)
    writer_dude = csv.DictWriter(out_dude, fieldnames=fieldnames)
    writer_dekois = csv.DictWriter(out_dekois, fieldnames=fieldnames)
    
    writer_cleaned.writeheader()
    writer_dude.writeheader()
    writer_dekois.writeheader()
    
    for row in reader:
        stats['total'] += 1
        if stats['total'] % 50000 == 0:
            print(f"   Processed {stats['total']:,} rows...")
        
        protein = row.get('Protein', '').strip()
        smiles = row.get('SMILES', '').strip()
        
        is_dude_protein = protein in dude_proteins
        is_dude_smiles = smiles in dude_smiles
        is_dekois_protein = protein in dekois_proteins
        is_dekois_smiles = smiles in dekois_smiles
        is_test_protein = protein in all_test_proteins
        is_test_smiles = smiles in all_test_smiles
        
        dude_overlap = is_dude_protein or is_dude_smiles
        dekois_overlap = is_dekois_protein or is_dekois_smiles
        test_overlap = is_test_protein or is_test_smiles
        
        if is_dude_protein:
            stats['dude_protein_overlap'] += 1
        if is_dude_smiles:
            stats['dude_smiles_overlap'] += 1
        if is_dude_protein and is_dude_smiles:
            stats['dude_both_overlap'] += 1
        
        if is_dekois_protein:
            stats['dekois_protein_overlap'] += 1
        if is_dekois_smiles:
            stats['dekois_smiles_overlap'] += 1
        if is_dekois_protein and is_dekois_smiles:
            stats['dekois_both_overlap'] += 1
        
        if is_test_protein:
            stats['test_protein_overlap'] += 1
        if is_test_smiles:
            stats['test_smiles_overlap'] += 1
        if is_test_protein and is_test_smiles:
            stats['test_both_overlap'] += 1
        
        if dude_overlap:
            writer_dude.writerow(row)
        
        if dekois_overlap:
            writer_dekois.writerow(row)
        
        if not test_overlap:
            writer_cleaned.writerow(row)
            stats['cleaned'] += 1

print(f"\n5. Generating report...")

report_lines = []
report_lines.append("=" * 80)
report_lines.append("Regression Dataset Test Set Overlap Removal Report")
report_lines.append(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
report_lines.append("=" * 80)
report_lines.append("")

report_lines.append("I. Input Data Statistics")
report_lines.append("-" * 80)
report_lines.append(f"Regression dataset: {REGRESSION_DATASET}")
report_lines.append(f"Total records: {stats['total']:,}")
report_lines.append("")

report_lines.append("II. Test Set Statistics")
report_lines.append("-" * 80)
report_lines.append(f"DUD-E:")
report_lines.append(f"  - Records: {dude_count:,}")
report_lines.append(f"  - Unique proteins: {len(dude_proteins):,}")
report_lines.append(f"  - Unique SMILES: {len(dude_smiles):,}")
report_lines.append("")
report_lines.append(f"DEKOIS2.0:")
report_lines.append(f"  - Records: {dekois_count:,}")
report_lines.append(f"  - Unique proteins: {len(dekois_proteins):,}")
report_lines.append(f"  - Unique SMILES: {len(dekois_smiles):,}")
report_lines.append("")
report_lines.append(f"Combined test set:")
report_lines.append(f"  - Total unique proteins: {len(all_test_proteins):,}")
report_lines.append(f"  - Total unique SMILES: {len(all_test_smiles):,}")
report_lines.append("")

report_lines.append("III. Overlap Statistics")
report_lines.append("-" * 80)
report_lines.append(f"DUD-E overlap:")
report_lines.append(f"  - Protein overlap: {stats['dude_protein_overlap']:,}")
report_lines.append(f"  - SMILES overlap: {stats['dude_smiles_overlap']:,}")
report_lines.append(f"  - Both overlap: {stats['dude_both_overlap']:,}")
dude_total_overlap = stats['dude_protein_overlap'] + stats['dude_smiles_overlap'] - stats['dude_both_overlap']
report_lines.append(f"  - Total overlap records: {dude_total_overlap:,}")
report_lines.append("")

report_lines.append(f"DEKOIS2.0 overlap:")
report_lines.append(f"  - Protein overlap: {stats['dekois_protein_overlap']:,}")
report_lines.append(f"  - SMILES overlap: {stats['dekois_smiles_overlap']:,}")
report_lines.append(f"  - Both overlap: {stats['dekois_both_overlap']:,}")
dekois_total_overlap = stats['dekois_protein_overlap'] + stats['dekois_smiles_overlap'] - stats['dekois_both_overlap']
report_lines.append(f"  - Total overlap records: {dekois_total_overlap:,}")
report_lines.append("")

report_lines.append(f"Overall test set overlap:")
report_lines.append(f"  - Protein overlap: {stats['test_protein_overlap']:,}")
report_lines.append(f"  - SMILES overlap: {stats['test_smiles_overlap']:,}")
report_lines.append(f"  - Both overlap: {stats['test_both_overlap']:,}")
test_total_overlap = stats['test_protein_overlap'] + stats['test_smiles_overlap'] - stats['test_both_overlap']
report_lines.append(f"  - Total overlap records: {test_total_overlap:,}")
report_lines.append("")

report_lines.append("IV. Processing Results")
report_lines.append("-" * 80)
report_lines.append(f"Cleaned dataset: {stats['cleaned']:,} ({stats['cleaned']/stats['total']*100:.2f}%)")
report_lines.append(f"Removed data: {stats['total'] - stats['cleaned']:,} ({(stats['total']-stats['cleaned'])/stats['total']*100:.2f}%)")
report_lines.append("")

report_lines.append("V. Output Files")
report_lines.append("-" * 80)
report_lines.append(f"1. Cleaned regression dataset: {OUTPUT_CLEANED}")
report_lines.append(f"2. DUD-E overlap data: {OUTPUT_DUDE_OVERLAP}")
report_lines.append(f"3. DEKOIS2.0 overlap data: {OUTPUT_DEKOIS_OVERLAP}")
report_lines.append("")

report_lines.append("=" * 80)

with open(REPORT_FILE, 'w', encoding='utf-8') as f:
    f.write('\n'.join(report_lines))

print(f"\nProcessing completed!")
print(f"=" * 70)
print(f"Total processed rows: {stats['total']:,}")
print(f"\nCleaned dataset: {stats['cleaned']:,} ({stats['cleaned']/stats['total']*100:.2f}%)")
print(f"Removed data: {stats['total'] - stats['cleaned']:,} ({(stats['total']-stats['cleaned'])/stats['total']*100:.2f}%)")

print(f"\nOverlap statistics:")
print(f"DUD-E overlap:")
print(f"  - Protein overlap: {stats['dude_protein_overlap']:,}")
print(f"  - SMILES overlap: {stats['dude_smiles_overlap']:,}")
print(f"  - Both overlap: {stats['dude_both_overlap']:,}")
print(f"  - Total overlap records: {dude_total_overlap:,}")

print(f"\nDEKOIS2.0 overlap:")
print(f"  - Protein overlap: {stats['dekois_protein_overlap']:,}")
print(f"  - SMILES overlap: {stats['dekois_smiles_overlap']:,}")
print(f"  - Both overlap: {stats['dekois_both_overlap']:,}")
print(f"  - Total overlap records: {dekois_total_overlap:,}")

print(f"\nOverall test set overlap:")
print(f"  - Protein overlap: {stats['test_protein_overlap']:,}")
print(f"  - SMILES overlap: {stats['test_smiles_overlap']:,}")
print(f"  - Both overlap: {stats['test_both_overlap']:,}")
print(f"  - Total overlap records: {test_total_overlap:,}")

print(f"\nOutput files:")
print(f"  1. Cleaned regression dataset: {OUTPUT_CLEANED}")
print(f"  2. DUD-E overlap data: {OUTPUT_DUDE_OVERLAP}")
print(f"  3. DEKOIS2.0 overlap data: {OUTPUT_DEKOIS_OVERLAP}")
print(f"  4. Processing report: {REPORT_FILE}")

print()
print("=" * 80)
```

```python

"""
Cluster protein sequences in step11_regression_dataset_cleaned.csv
Using cd-hit tool, 70% similarity
Add target_cluster column to each entry
"""

import csv
import subprocess
import re
from pathlib import Path

BASE_DIR = Path.cwd()
INPUT_FILE = BASE_DIR / "step11_regression_dataset_cleaned.csv"
OUTPUT_FILE = BASE_DIR / "step12_regression_dataset_with_clusters.csv"

FASTA_FILE = BASE_DIR / "proteins_for_clustering_reg.fasta"
CDHIT_OUTPUT = BASE_DIR / "proteins_clustered_reg"
CDHIT_CLSTR = BASE_DIR / "proteins_clustered_reg.clstr"

print("=" * 80)
print("Protein Sequence Clustering (Regression Dataset)")
print("=" * 80)
print()

print("1. Extracting all unique protein sequences...")
unique_proteins = {}
protein_index = 0

with open(INPUT_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        protein = row.get('Protein', '').strip()
        if protein and protein not in unique_proteins:
            unique_proteins[protein] = protein_index
            protein_index += 1

print(f"   Found {len(unique_proteins):,} unique protein sequences")
print()

print("2. Writing sequences to FASTA file...")
with open(FASTA_FILE, 'w', encoding='utf-8') as f:
    for protein, idx in unique_proteins.items():
        f.write(f">seq_{idx}\n")
        for i in range(0, len(protein), 80):
            f.write(protein[i:i+80] + "\n")

print(f"   Written {len(unique_proteins):,} sequences to {FASTA_FILE}")
print()

print("3. Running cd-hit for clustering (70% similarity)...")
print("   Command: cd-hit -i proteins_for_clustering_reg.fasta -o proteins_clustered_reg -c 0.7 -n 5 -T 0 -M 0")
print()

try:
    result = subprocess.run(['cd-hit', '-h'], capture_output=True, text=True, timeout=5)
    cdhit_available = True
except (FileNotFoundError, subprocess.TimeoutExpired):
    cdhit_available = False
    print("   Error: cd-hit not found, please ensure cd-hit is installed and in PATH")
    print("   Installation: conda install -c bioconda cd-hit")
    exit(1)

cmd = [
    'cd-hit',
    '-i', str(FASTA_FILE),
    '-o', str(CDHIT_OUTPUT),
    '-c', '0.7',
    '-n', '5',
    '-T', '0',
    '-M', '0'
]

print("   Running cd-hit (this may take some time)...")
try:
    result = subprocess.run(cmd, capture_output=True, text=True, check=True)
    print("   cd-hit completed")
    if result.stderr:
        lines = result.stderr.split('\n')
        for line in lines:
            if any(keyword in line.lower() for keyword in ['total seq', 'clusters', 'longest', 'shortest']):
                print(f"   {line.strip()}")
except subprocess.CalledProcessError as e:
    print(f"   Error: cd-hit failed")
    print(f"   Error message: {e.stderr}")
    exit(1)

print()

print("4. Parsing clustering results...")

if not CDHIT_CLSTR.exists():
    print(f"   Error: Clustering result file does not exist: {CDHIT_CLSTR}")
    exit(1)

sequence_to_cluster = {}
current_cluster = None
cluster_count = 0

with open(CDHIT_CLSTR, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line.startswith('>Cluster'):
            match = re.search(r'>Cluster\s+(\d+)', line)
            if match:
                current_cluster = int(match.group(1))
                cluster_count = max(cluster_count, current_cluster + 1)
        elif line and current_cluster is not None:
            match = re.search(r'>seq_(\d+)', line)
            if match:
                seq_idx = int(match.group(1))
                sequence_to_cluster[seq_idx] = current_cluster

print(f"   Clustered into {cluster_count:,} clusters")
print(f"   Parsed {len(sequence_to_cluster):,} sequence cluster assignments")
print()

print("5. Building protein sequence to cluster mapping...")
protein_to_cluster = {}

index_to_protein = {idx: seq for seq, idx in unique_proteins.items()}

for seq_idx, cluster_id in sequence_to_cluster.items():
    if seq_idx in index_to_protein:
        protein = index_to_protein[seq_idx]
        protein_to_cluster[protein] = cluster_id

unassigned_count = 0
for protein in unique_proteins:
    if protein not in protein_to_cluster:
        protein_to_cluster[protein] = cluster_count
        cluster_count += 1
        unassigned_count += 1

if unassigned_count > 0:
    print(f"   Warning: {unassigned_count} sequences not found in clustering results, assigned new clusters")

print(f"   Built {len(protein_to_cluster):,} sequence mappings")
print()

print("6. Adding target_cluster column to original data...")

stats = {
    'total': 0,
    'with_cluster': 0,
    'without_cluster': 0
}

with open(INPUT_FILE, 'r', encoding='utf-8') as infile, \
     open(OUTPUT_FILE, 'w', encoding='utf-8', newline='') as outfile:
    
    reader = csv.DictReader(infile)
    fieldnames = list(reader.fieldnames)
    
    if 'target_cluster' in fieldnames:
        fieldnames.remove('target_cluster')
    
    fieldnames.append('target_cluster')
    
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()
    
    for row in reader:
        stats['total'] += 1
        
        protein = row.get('Protein', '').strip()
        cluster_id = protein_to_cluster.get(protein)
        
        if cluster_id is not None:
            row['target_cluster'] = cluster_id
            stats['with_cluster'] += 1
        else:
            row['target_cluster'] = cluster_count
            stats['without_cluster'] += 1
        
        writer.writerow(row)
        
        if stats['total'] % 100000 == 0:
            print(f"   Processed {stats['total']:,} rows...")

print()

print("7. Statistics on cluster distribution...")
cluster_sizes = {}
with open(OUTPUT_FILE, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        cluster_id = row.get('target_cluster', '').strip()
        if cluster_id:
            try:
                cluster_id = int(cluster_id)
                cluster_sizes[cluster_id] = cluster_sizes.get(cluster_id, 0) + 1
            except ValueError:
                pass

print(f"   Number of clusters: {len(cluster_sizes):,}")
if cluster_sizes:
    sizes = sorted(cluster_sizes.values())
    print(f"   Minimum cluster size: {min(sizes):,}")
    print(f"   Maximum cluster size: {max(sizes):,}")
    print(f"   Average cluster size: {sum(sizes)/len(sizes):.2f}")
    print(f"   Median cluster size: {sizes[len(sizes)//2]:,}")

size_distribution = {}
for size in sizes:
    if size < 10:
        size_distribution['1-9'] = size_distribution.get('1-9', 0) + 1
    elif size < 100:
        size_distribution['10-99'] = size_distribution.get('10-99', 0) + 1
    elif size < 1000:
        size_distribution['100-999'] = size_distribution.get('100-999', 0) + 1
    else:
        size_distribution['1000+'] = size_distribution.get('1000+', 0) + 1

print(f"\n   Cluster size distribution:")
for size_range, count in sorted(size_distribution.items()):
    print(f"     - {size_range}: {count:,} clusters")

print()

print("=" * 80)
print("Clustering completed!")
print("=" * 80)
print(f"Input file: {INPUT_FILE}")
print(f"Output file: {OUTPUT_FILE}")
print(f"FASTA file: {FASTA_FILE}")
print(f"cd-hit output: {CDHIT_OUTPUT}")
print()
print(f"Statistics:")
print(f"  - Unique protein sequences: {len(unique_proteins):,}")
print(f"  - Clusters after clustering: {cluster_count:,}")
print(f"  - Total data records: {stats['total']:,}")
print(f"  - Records with clusters: {stats['with_cluster']:,}")
if stats['without_cluster'] > 0:
    print(f"  - Records without clusters: {stats['without_cluster']:,}")
print()
print(f"Clustering parameters: 70% similarity")
print("=" * 80)
```

```python
"""
Data augmentation for regression dataset
For high-affinity samples (Y > 7), find high-affinity samples (Y > 7) from other clusters
with fingerprint similarity < 0.3 to all high-affinity samples in the original cluster.
Generate negative samples: keep original protein, replace SMILES, Y = original_Y - 4
"""

import pandas as pd
import numpy as np
from tqdm import tqdm
from rdkit import Chem
from rdkit import DataStructs
from rdkit.Chem import rdFingerprintGenerator
import argparse
import os
from typing import Optional, Dict, List, Tuple
from pathlib import Path

MORGAN_GENERATOR = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048, includeChirality=False)

def calculate_fingerprint(smiles: str) -> Optional[object]:
    """
    Calculate 2048-bit binary fingerprint using Morgan algorithm (radius=2)
    """
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None
        fingerprint = MORGAN_GENERATOR.GetFingerprint(mol)
        return fingerprint
    except Exception as e:
        print(f"[FP ERROR] SMILES = {smiles}, error = {e}")
        return None


def precompute_fingerprints(df: pd.DataFrame, smiles_col: str = "SMILES") -> List[Optional[object]]:
    """
    Precompute fingerprints for all rows in df, returns a list aligned with df.index
    """
    fps: List[Optional[object]] = []
    print("Precomputing Morgan fingerprints for all compounds...")
    for smi in tqdm(df[smiles_col].tolist(), desc="Fingerprints"):
        fps.append(calculate_fingerprint(smi))
    valid = sum(fp is not None for fp in fps)
    print(f"  - Total compounds: {len(fps)}, valid fingerprints: {valid}")
    return fps


def build_cluster_index(df_high_affinity: pd.DataFrame) -> Dict[int, List[int]]:
    """
    Build {cluster_id: [row_index_list]} mapping based on target_cluster.
    Uses integer index from df_high_affinity (after reset_index).
    """
    cluster_ids = df_high_affinity["target_cluster"].values
    cluster_index: Dict[int, List[int]] = {}
    for idx, c in enumerate(cluster_ids):
        cluster_index.setdefault(c, []).append(idx)
    return cluster_index


def is_candidate_dissimilar_to_cluster(
    cand_fp: object,
    cluster_high_affinity_indices: List[int],
    fps_high_affinity: List[Optional[object]],
    threshold: float
) -> bool:
    """
    Check if candidate ligand fingerprint satisfies Tanimoto(candidate, each_high_affinity) < threshold
    for all high-affinity samples in the cluster.
    """
    if cand_fp is None:
        return False

    for idx in cluster_high_affinity_indices:
        fp_ha = fps_high_affinity[idx]
        if fp_ha is None:
            continue
        sim = DataStructs.TanimotoSimilarity(cand_fp, fp_ha)
        if sim >= threshold:
            return False
    return True


def generate_negatives_for_cluster(
    cluster_id: int,
    cluster_high_affinity_indices: List[int],
    df_high_affinity: pd.DataFrame,
    fps_high_affinity: List[Optional[object]],
    global_used_candidate_indices: set,
    similarity_threshold: float,
    max_trials_factor: float = 10.0
) -> List[Tuple[int, int]]:
    """
    Generate negative samples for a cluster. For each high-affinity sample in the cluster,
    select a candidate ligand from other clusters, requiring:
      - Candidate belongs to other clusters
      - Candidate is high-affinity (Y > 7)
      - Candidate not used before (globally unique)
      - Candidate Tanimoto similarity to all high-affinity samples in this cluster < threshold
    Returns list of [(ha_idx, cand_idx), ...], length ideally equals cluster high-affinity count.
    """
    n_ha = len(cluster_high_affinity_indices)
    if n_ha == 0:
        return []

    all_indices = np.arange(len(df_high_affinity))
    cluster_arr = df_high_affinity["target_cluster"].values

    candidate_mask = (cluster_arr != cluster_id)
    candidate_indices = all_indices[candidate_mask]

    candidate_indices = [
        idx for idx in candidate_indices
        if (idx not in global_used_candidate_indices) and (fps_high_affinity[idx] is not None)
    ]

    if len(candidate_indices) == 0:
        print(f"[WARN] Cluster {cluster_id}: no available candidate ligands from other clusters.")
        return []

    rng = np.random.default_rng()
    rng.shuffle(candidate_indices)

    max_trials = int(max_trials_factor * n_ha)
    if max_trials > len(candidate_indices):
        max_trials = len(candidate_indices)

    selected_pairs: List[Tuple[int, int]] = []
    used_in_this_cluster: set = set()

    cluster_ha_indices_local = list(cluster_high_affinity_indices)

    cand_ptr = 0
    good_candidates: List[int] = []

    while len(good_candidates) < n_ha and cand_ptr < max_trials:
        cand_idx = candidate_indices[cand_ptr]
        cand_ptr += 1

        if cand_idx in global_used_candidate_indices or cand_idx in used_in_this_cluster:
            continue

        cand_fp = fps_high_affinity[cand_idx]
        if cand_fp is None:
            continue

        if is_candidate_dissimilar_to_cluster(
            cand_fp,
            cluster_ha_indices_local,
            fps_high_affinity,
            similarity_threshold
        ):
            good_candidates.append(cand_idx)
            used_in_this_cluster.add(cand_idx)
            global_used_candidate_indices.add(cand_idx)

    if len(good_candidates) == 0:
        print(f"[WARN] Cluster {cluster_id}: no dissimilar candidates found (under threshold={similarity_threshold}).")
        return []

    if len(good_candidates) < n_ha:
        print(
            f"[WARN] Cluster {cluster_id}: only found {len(good_candidates)} dissimilar candidates "
            f"for {n_ha} high-affinity samples. Will only augment the first {len(good_candidates)} samples."
        )

    rng.shuffle(cluster_ha_indices_local)
    pair_count = min(len(cluster_ha_indices_local), len(good_candidates))
    for i in range(pair_count):
        ha_idx = cluster_ha_indices_local[i]
        cand_idx = good_candidates[i]
        selected_pairs.append((ha_idx, cand_idx))

    return selected_pairs


def augment_regression_data(
    input_file: str,
    output_file: str,
    high_affinity_threshold: float = 7.0,
    similarity_threshold: float = 0.3,
    y_decrement: float = 4.0,
    random_seed: int = 0
) -> None:
    """
    Data augmentation for regression dataset:
      - Based on target_cluster (from CD-HIT clustering)
      - Process at cluster level
      - All high-affinity samples (Y > threshold) in a cluster treated as actives
      - Negatives from high-affinity samples (Y > threshold) in other clusters
      - Candidate must have Tanimoto similarity < threshold to all high-affinity samples in cluster
      - Each candidate used only once
      - Skip sample if no suitable candidate found
    
    Generated negative sample rows:
      - Copy a high-affinity row from the cluster
      - Keep Protein/target_cluster target information
      - Replace SMILES with ligand from other cluster
      - Set Y = original_Y - y_decrement
      - Set standard_value and standard_units to empty
    """
    np.random.seed(random_seed)

    print(f"Reading data from {input_file} ...")
    df = pd.read_csv(input_file)

    if "Y" not in df.columns:
        raise ValueError("Input CSV must contain column 'Y' as affinity value.")
    if "SMILES" not in df.columns:
        raise ValueError("Input CSV must contain column 'SMILES'.")
    if "target_cluster" not in df.columns:
        raise ValueError("Input CSV must contain column 'target_cluster'.")
    if "Protein" not in df.columns:
        raise ValueError("Input CSV must contain column 'Protein'.")

    df_high_affinity = df[df["Y"] > high_affinity_threshold].copy()
    df_high_affinity.reset_index(drop=True, inplace=True)

    print(f"Total high-affinity samples (Y > {high_affinity_threshold}): {len(df_high_affinity)}")

    fps_high_affinity = precompute_fingerprints(df_high_affinity, smiles_col="SMILES")

    cluster_index = build_cluster_index(df_high_affinity)
    all_clusters = sorted(cluster_index.keys())
    print(f"Number of target clusters (with high-affinity samples): {len(all_clusters)}")

    global_used_candidate_indices: set = set()

    negative_rows = []

    for cluster_id in tqdm(all_clusters, desc="Processing clusters"):
        ha_indices = cluster_index[cluster_id]
        if len(ha_indices) == 0:
            continue

        pairs = generate_negatives_for_cluster(
            cluster_id=cluster_id,
            cluster_high_affinity_indices=ha_indices,
            df_high_affinity=df_high_affinity,
            fps_high_affinity=fps_high_affinity,
            global_used_candidate_indices=global_used_candidate_indices,
            similarity_threshold=similarity_threshold,
            max_trials_factor=10.0,
        )

        for ha_idx, cand_idx in pairs:
            ha_row = df_high_affinity.loc[ha_idx]
            cand_row = df_high_affinity.loc[cand_idx]

            new_row = ha_row.copy()
            new_row["SMILES"] = cand_row["SMILES"]
            original_y = float(ha_row["Y"])
            new_row["Y"] = original_y - y_decrement
            
            if "standard_value" in new_row:
                new_row["standard_value"] = ""
            if "standard_units" in new_row:
                new_row["standard_units"] = ""

            negative_rows.append(new_row)

    if len(negative_rows) == 0:
        print("[WARN] No negative samples were generated. Please check your threshold or data.")
        df_aug = df
    else:
        df_negative = pd.DataFrame(negative_rows)
        print(f"Generated negative samples: {len(df_negative)}")

        df_aug = pd.concat([df, df_negative], ignore_index=True)

    print(f"Saving augmented dataset to {output_file} ...")
    df_aug.to_csv(output_file, index=False)
    print("Done.")
    print(f"Original samples: {len(df)}")
    print(f"Augmented samples: {len(df_aug)}")
    print(f"New negative samples: {len(negative_rows)}")


def main():
    BASE_DIR = Path(__file__).parent
    parser = argparse.ArgumentParser(
        description="Data augmentation for regression dataset (cluster-wise conditional label decrement)."
    )
    parser.add_argument("--input", "-i", default=str(BASE_DIR / "step12_regression_dataset_with_clusters.csv"), help="Input CSV file path (with target_cluster column)")
    parser.add_argument("--output", "-o", default=str(BASE_DIR / "step13_regression_dataset_with_negatives.csv"), help="Output CSV file path")
    parser.add_argument(
        "--high-affinity-threshold", "-ha", type=float, default=7.0,
        help="High affinity threshold (default: 7.0, i.e., Y > 7)"
    )
    parser.add_argument(
        "--threshold", "-t", type=float, default=0.3,
        help="Tanimoto similarity threshold for negative sampling (default: 0.3)"
    )
    parser.add_argument(
        "--y-decrement", "-d", type=float, default=4.0,
        help="Y value decrement for negative samples (default: 4.0, i.e., Y_new = Y_original - 4)"
    )
    parser.add_argument("--seed", "-s", type=int, default=0, help="Random seed (default: 0)")

    args = parser.parse_args()

    if not os.path.exists(args.input):
        print(f"Error: input file {args.input} not found!")
        return

    out_dir = os.path.dirname(args.output)
    if out_dir and not os.path.exists(out_dir):
        os.makedirs(out_dir)

    augment_regression_data(
        input_file=args.input,
        output_file=args.output,
        high_affinity_threshold=args.high_affinity_threshold,
        similarity_threshold=args.threshold,
        y_decrement=args.y_decrement,
        random_seed=args.seed
    )


if __name__ == "__main__":
    BASE_DIR = Path.cwd()
    INPUT_FILE = BASE_DIR / "step12_regression_dataset_with_clusters.csv"
    OUTPUT_FILE = BASE_DIR / "step13_regression_dataset_with_negatives.csv"
    
    if not INPUT_FILE.exists():
        print(f"Error: Input file not found: {INPUT_FILE}")
        print("Please ensure you are running this script in the directory containing step12_regression_dataset_with_clusters.csv")
        exit(1)
    
    augment_regression_data(
        input_file=str(INPUT_FILE),
        output_file=str(OUTPUT_FILE),
        high_affinity_threshold=7.0,
        similarity_threshold=0.3,
        y_decrement=4.0,
        random_seed=0
    )
```

```python
"""
Statistics script: Count molecules that have both positive and negative samples
- Analyze augmented dataset (step9_final_dataset.csv)
- Analyze unaugmented dataset (step7_classification_dataset_with_clusters.csv)
- Analyze regression datasets (targets and unique molecules)
"""

import csv
from pathlib import Path
from collections import defaultdict
from datetime import datetime

BASE_DIR = Path.cwd()

AUGMENTED_FILE = BASE_DIR / "step9_final_dataset.csv"
UNAUGMENTED_FILE = BASE_DIR / "step7_classification_dataset_with_clusters.csv"
REGRESSION_UNAUGMENTED_FILE = BASE_DIR / "step12_regression_dataset_with_clusters.csv"
REGRESSION_AUGMENTED_FILE = BASE_DIR / "step13_regression_dataset_with_negatives.csv"
REPORT_FILE = BASE_DIR / f"statistics_molecules_with_both_labels_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"

print("=" * 80)
print("Statistics: Count molecules with both positive and negative samples")
print("=" * 80)
print()

def analyze_dataset(file_path, dataset_name, is_regression=False):
    """
    Analyze dataset and count molecules with both positive and negative samples
    
    Args:
        file_path: Path to CSV file
        dataset_name: Dataset name for display
        is_regression: Whether this is a regression dataset (only count targets and molecules, not labels)
    
    Returns:
        dict: Statistics results
    """
    print(f"Analyzing: {dataset_name}")
    print(f"File path: {file_path}")
    
    if not file_path.exists():
        print(f"Error: File not found: {file_path}")
        return None
    
    smiles_stats = defaultdict(lambda: {'positive': 0, 'negative': 0, 'total': 0})
    unique_targets = set()
    total_rows = 0
    valid_rows = 0
    
    with open(file_path, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        
        required_cols = ['SMILES', 'Y']
        if 'Protein' in reader.fieldnames:
            required_cols.append('Protein')
        missing_cols = [col for col in required_cols if col not in reader.fieldnames]
        if missing_cols:
            print(f"Error: Missing required columns: {', '.join(missing_cols)}")
            print(f"Available columns: {', '.join(reader.fieldnames)}")
            return None
        
        for row in reader:
            total_rows += 1
            
            if total_rows % 100000 == 0:
                print(f"  Processed {total_rows:,} rows...")
            
            smiles = row.get('SMILES', '').strip()
            y_str = row.get('Y', '').strip()
            protein = row.get('Protein', '').strip()
            
            if not smiles or not y_str:
                continue
            
            if protein:
                unique_targets.add(protein)
            
            if is_regression:
                try:
                    float(y_str)
                    valid_rows += 1
                    smiles_stats[smiles]['total'] += 1
                except (ValueError, TypeError):
                    continue
            else:
                try:
                    y = int(y_str)
                    valid_rows += 1
                    smiles_stats[smiles]['total'] += 1
                    
                    if y == 1:
                        smiles_stats[smiles]['positive'] += 1
                    elif y == 0:
                        smiles_stats[smiles]['negative'] += 1
                except (ValueError, TypeError):
                    continue
    
    print(f"  Total rows: {total_rows:,}")
    print(f"  Valid rows: {valid_rows:,}")
    print(f"  Unique SMILES: {len(smiles_stats):,}")
    print(f"  Unique Targets: {len(unique_targets):,}")
    print()
    
    if is_regression:
        result = {
            'dataset_name': dataset_name,
            'file_path': str(file_path),
            'total_rows': total_rows,
            'valid_rows': valid_rows,
            'unique_smiles': len(smiles_stats),
            'unique_targets': len(unique_targets),
            'is_regression': True
        }
        return result
    
    molecules_with_both = []
    only_positive = 0
    only_negative = 0
    both_labels = 0
    
    for smiles, stats in smiles_stats.items():
        has_positive = stats['positive'] > 0
        has_negative = stats['negative'] > 0
        
        if has_positive and has_negative:
            both_labels += 1
            molecules_with_both.append({
                'SMILES': smiles,
                'positive_count': stats['positive'],
                'negative_count': stats['negative'],
                'total_count': stats['total']
            })
        elif has_positive:
            only_positive += 1
        elif has_negative:
            only_negative += 1
    
    result = {
        'dataset_name': dataset_name,
        'file_path': str(file_path),
        'total_rows': total_rows,
        'valid_rows': valid_rows,
        'unique_smiles': len(smiles_stats),
        'unique_targets': len(unique_targets),
        'only_positive': only_positive,
        'only_negative': only_negative,
        'both_labels': both_labels,
        'molecules_with_both': molecules_with_both,
        'is_regression': False
    }
    
    return result

def print_statistics(result):
    """Print statistics results"""
    if result is None:
        return
    
    print(f"Dataset: {result['dataset_name']}")
    print("-" * 80)
    print(f"Total rows: {result['total_rows']:,}")
    print(f"Valid rows: {result['valid_rows']:,}")
    print(f"Unique SMILES: {result['unique_smiles']:,}")
    print(f"Unique Targets: {result['unique_targets']:,}")
    print()
    
    if result.get('is_regression', False):
        return
    
    print("Label distribution:")
    print(f"  Only positive (Y=1): {result['only_positive']:,} ({result['only_positive']/result['unique_smiles']*100:.2f}%)")
    print(f"  Only negative (Y=0): {result['only_negative']:,} ({result['only_negative']/result['unique_smiles']*100:.2f}%)")
    print(f"  Both labels: {result['both_labels']:,} ({result['both_labels']/result['unique_smiles']*100:.2f}%)")
    print()
    
    if result['both_labels'] > 0:
        total_positive_in_both = sum(m['positive_count'] for m in result['molecules_with_both'])
        total_negative_in_both = sum(m['negative_count'] for m in result['molecules_with_both'])
        avg_positive = total_positive_in_both / result['both_labels']
        avg_negative = total_negative_in_both / result['both_labels']
        
        print("Details for molecules with both labels:")
        print(f"  Total positive samples: {total_positive_in_both:,}")
        print(f"  Total negative samples: {total_negative_in_both:,}")
        print(f"  Average positive samples per molecule: {avg_positive:.2f}")
        print(f"  Average negative samples per molecule: {avg_negative:.2f}")
        print()
        
        print("Top 10 examples of molecules with both labels:")
        sorted_examples = sorted(result['molecules_with_both'], 
                                key=lambda x: x['total_count'], 
                                reverse=True)[:10]
        for i, mol in enumerate(sorted_examples, 1):
            print(f"  {i}. SMILES: {mol['SMILES'][:80]}...")
            print(f"     Positive: {mol['positive_count']}, Negative: {mol['negative_count']}, Total: {mol['total_count']}")
    print()

print("1. Analyzing augmented dataset (step9_final_dataset.csv)")
print("=" * 80)
result_augmented = analyze_dataset(AUGMENTED_FILE, "Augmented dataset (step9_final_dataset.csv)")
print_statistics(result_augmented)
print()

print("2. Analyzing unaugmented dataset (step7_classification_dataset_with_clusters.csv)")
print("=" * 80)
result_unaugmented = analyze_dataset(UNAUGMENTED_FILE, "Unaugmented dataset (step7_classification_dataset_with_clusters.csv)")
print_statistics(result_unaugmented)
print()

print("3. Analyzing regression unaugmented dataset (step12_regression_dataset_with_clusters.csv)")
print("=" * 80)
result_regression_unaugmented = analyze_dataset(REGRESSION_UNAUGMENTED_FILE, "Regression unaugmented dataset (step12_regression_dataset_with_clusters.csv)", is_regression=True)
print_statistics(result_regression_unaugmented)
print()

print("4. Analyzing regression augmented dataset (step13_regression_dataset_with_negatives.csv)")
print("=" * 80)
result_regression_augmented = analyze_dataset(REGRESSION_AUGMENTED_FILE, "Regression augmented dataset (step13_regression_dataset_with_negatives.csv)", is_regression=True)
print_statistics(result_regression_augmented)
print()

print("5. Generating statistics report...")
report_lines = []
report_lines.append("=" * 80)
report_lines.append("Molecule Label Statistics Report")
report_lines.append(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
report_lines.append("=" * 80)
report_lines.append("")

if result_augmented:
    report_lines.append("I. Augmented Dataset (step9_final_dataset.csv)")
    report_lines.append("-" * 80)
    report_lines.append(f"File path: {result_augmented['file_path']}")
    report_lines.append(f"Total rows: {result_augmented['total_rows']:,}")
    report_lines.append(f"Valid rows: {result_augmented['valid_rows']:,}")
    report_lines.append(f"Unique SMILES: {result_augmented['unique_smiles']:,}")
    report_lines.append(f"Unique Targets: {result_augmented['unique_targets']:,}")
    report_lines.append("")
    report_lines.append("Label distribution:")
    report_lines.append(f"  Only positive (Y=1): {result_augmented['only_positive']:,} ({result_augmented['only_positive']/result_augmented['unique_smiles']*100:.2f}%)")
    report_lines.append(f"  Only negative (Y=0): {result_augmented['only_negative']:,} ({result_augmented['only_negative']/result_augmented['unique_smiles']*100:.2f}%)")
    report_lines.append(f"  Both labels: {result_augmented['both_labels']:,} ({result_augmented['both_labels']/result_augmented['unique_smiles']*100:.2f}%)")
    report_lines.append("")
    
    if result_augmented['both_labels'] > 0:
        total_positive_in_both = sum(m['positive_count'] for m in result_augmented['molecules_with_both'])
        total_negative_in_both = sum(m['negative_count'] for m in result_augmented['molecules_with_both'])
        avg_positive = total_positive_in_both / result_augmented['both_labels']
        avg_negative = total_negative_in_both / result_augmented['both_labels']
        
        report_lines.append("Details for molecules with both labels:")
        report_lines.append(f"  Total positive samples: {total_positive_in_both:,}")
        report_lines.append(f"  Total negative samples: {total_negative_in_both:,}")
        report_lines.append(f"  Average positive samples per molecule: {avg_positive:.2f}")
        report_lines.append(f"  Average negative samples per molecule: {avg_negative:.2f}")
    report_lines.append("")

if result_unaugmented:
    report_lines.append("II. Unaugmented Dataset (step7_classification_dataset_with_clusters.csv)")
    report_lines.append("-" * 80)
    report_lines.append(f"File path: {result_unaugmented['file_path']}")
    report_lines.append(f"Total rows: {result_unaugmented['total_rows']:,}")
    report_lines.append(f"Valid rows: {result_unaugmented['valid_rows']:,}")
    report_lines.append(f"Unique SMILES: {result_unaugmented['unique_smiles']:,}")
    report_lines.append(f"Unique Targets: {result_unaugmented['unique_targets']:,}")
    report_lines.append("")
    report_lines.append("Label distribution:")
    report_lines.append(f"  Only positive (Y=1): {result_unaugmented['only_positive']:,} ({result_unaugmented['only_positive']/result_unaugmented['unique_smiles']*100:.2f}%)")
    report_lines.append(f"  Only negative (Y=0): {result_unaugmented['only_negative']:,} ({result_unaugmented['only_negative']/result_unaugmented['unique_smiles']*100:.2f}%)")
    report_lines.append(f"  Both labels: {result_unaugmented['both_labels']:,} ({result_unaugmented['both_labels']/result_unaugmented['unique_smiles']*100:.2f}%)")
    report_lines.append("")
    
    if result_unaugmented['both_labels'] > 0:
        total_positive_in_both = sum(m['positive_count'] for m in result_unaugmented['molecules_with_both'])
        total_negative_in_both = sum(m['negative_count'] for m in result_unaugmented['molecules_with_both'])
        avg_positive = total_positive_in_both / result_unaugmented['both_labels']
        avg_negative = total_negative_in_both / result_unaugmented['both_labels']
        
        report_lines.append("Details for molecules with both labels:")
        report_lines.append(f"  Total positive samples: {total_positive_in_both:,}")
        report_lines.append(f"  Total negative samples: {total_negative_in_both:,}")
        report_lines.append(f"  Average positive samples per molecule: {avg_positive:.2f}")
        report_lines.append(f"  Average negative samples per molecule: {avg_negative:.2f}")
    report_lines.append("")

if result_regression_unaugmented:
    report_lines.append("III. Regression Unaugmented Dataset (step12_regression_dataset_with_clusters.csv)")
    report_lines.append("-" * 80)
    report_lines.append(f"File path: {result_regression_unaugmented['file_path']}")
    report_lines.append(f"Total rows: {result_regression_unaugmented['total_rows']:,}")
    report_lines.append(f"Valid rows: {result_regression_unaugmented['valid_rows']:,}")
    report_lines.append(f"Unique SMILES: {result_regression_unaugmented['unique_smiles']:,}")
    report_lines.append(f"Unique Targets: {result_regression_unaugmented['unique_targets']:,}")
    report_lines.append("")

if result_regression_augmented:
    report_lines.append("IV. Regression Augmented Dataset (step13_regression_dataset_with_negatives.csv)")
    report_lines.append("-" * 80)
    report_lines.append(f"File path: {result_regression_augmented['file_path']}")
    report_lines.append(f"Total rows: {result_regression_augmented['total_rows']:,}")
    report_lines.append(f"Valid rows: {result_regression_augmented['valid_rows']:,}")
    report_lines.append(f"Unique SMILES: {result_regression_augmented['unique_smiles']:,}")
    report_lines.append(f"Unique Targets: {result_regression_augmented['unique_targets']:,}")
    report_lines.append("")

if result_augmented and result_unaugmented:
    report_lines.append("V. Classification Dataset Comparison")
    report_lines.append("-" * 80)
    report_lines.append(f"Molecules with both labels (augmented): {result_augmented['both_labels']:,}")
    report_lines.append(f"Molecules with both labels (unaugmented): {result_unaugmented['both_labels']:,}")
    diff = result_augmented['both_labels'] - result_unaugmented['both_labels']
    report_lines.append(f"Difference: {diff:,} ({diff/result_unaugmented['both_labels']*100:.2f}%)" if result_unaugmented['both_labels'] > 0 else "Difference: N/A")
    report_lines.append("")

if result_regression_augmented and result_regression_unaugmented:
    report_lines.append("VI. Regression Dataset Comparison")
    report_lines.append("-" * 80)
    report_lines.append(f"Unique SMILES (augmented): {result_regression_augmented['unique_smiles']:,}")
    report_lines.append(f"Unique SMILES (unaugmented): {result_regression_unaugmented['unique_smiles']:,}")
    smiles_diff = result_regression_augmented['unique_smiles'] - result_regression_unaugmented['unique_smiles']
    report_lines.append(f"SMILES difference: {smiles_diff:,}")
    report_lines.append(f"Unique Targets (augmented): {result_regression_augmented['unique_targets']:,}")
    report_lines.append(f"Unique Targets (unaugmented): {result_regression_unaugmented['unique_targets']:,}")
    target_diff = result_regression_augmented['unique_targets'] - result_regression_unaugmented['unique_targets']
    report_lines.append(f"Target difference: {target_diff:,}")
    report_lines.append("")

report_lines.append("=" * 80)

with open(REPORT_FILE, 'w', encoding='utf-8') as f:
    f.write('\n'.join(report_lines))

print(f"Report saved to: {REPORT_FILE}")
print()
print("=" * 80)
print("Statistics completed!")
print("=" * 80)
```
